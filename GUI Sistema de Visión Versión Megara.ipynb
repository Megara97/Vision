```python
import tkinter as tk
import tkinter.font as tkFont
from tkinter import filedialog
from PIL import Image, ImageTk
import numpy as np
import cv2
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import os
import math
import pandas as pd
from sympy import symbols, solve
import datetime

class App:
    def __init__(self,root):
        root.title("Sistema de visión")
        w=1500
        h=750
        sw=root.winfo_screenwidth()
        sh=root.winfo_screenheight()
        #alineacion='%dx%d+%d+%d'%(w,h,0,0) #Esquina superior izquierda
        alineacion='%dx%d+%d+%d'%(w,h,(sw-w)/2,(sh-h)/2) #Centrado
        root.geometry(alineacion)
        root.resizable(width=False,height=False) #Para no permitir maximizar la pantalla
        #root.configure(background="#ffffff")
        self.create_widgets()

    def create_widgets(self):
        global vid,Ls,ROI1,ROI2,labpath,butmen,butmas,Gfoto,rbgfoto,labherramientas,Opciones,lbxMedicion,Medicion,Figura,mbopcionmenu,tmedicion,rbtmed,labtmed,rbfig,frame3,Mood,but0,abin,insp,j,directorio,im2,lbxherramienta,lbxinspeccion,img,canvas,labpos,ax2,canvas2,ax3,canvas3,Herramienta,px,py,ancho,alto,mw,labROIs,labaROI,labvalor,labtolmin,labtolmax,enttolmax,labumbral,entumb,labumbalto,entumbalto,frame1,frame2,tolmin,tolmax,modo,polaridad,umbral,umbalto,valor,Resultado,rbmodo,rbpol,labx,laby,labw,labh,labmw,entmw,labpolaridad,rbpol,Resultado2,labrresult2 
        j=0
        insp=[]
        abin=0
        Ls=[]
        
        canvas=tk.Canvas(root) #Imagen principal
        canvas["background"]="#000000"
        img=np.zeros([480,640])
        canvas.place(x=260,y=10,width=641,height=481)
        canvas.bind('<Button-1>',self.ROIxy)
        canvas.bind('<Motion>',self.motion)
        canvas.bind('<ButtonRelease-1>',self.ROIwh)
        canvas.bind('<Button-3>',self.ROIrmin) #Double-Button-1
        
        labpos=tk.Label(root) #Posición del raton en la imagen
        #labpos["background"]="#000000"
        #labpos["foreground"]="#ffffff"
        labpos["font"]=tkFont.Font(family='FreeSans',size=7)
        labpos["anchor"]= "center"
        labpos["relief"]="flat"
        labpos.place(x=530,y=490,width=100,height=20) 
        
        parameters = {'axes.titlesize': 7,'xtick.labelsize': 7,'ytick.labelsize': 7}
        plt.rcParams.update(parameters)      
        fig2=plt.figure() #figsize=(4,2)
        ax2=fig2.gca()
        ax2.axis('on')
        ax2.set_title("Histograma de la ROI",fontsize=10)
        ax2.set_xlabel("Escala de grises")
        ax2.set_ylabel("Pixeles")
        ax2.set_xlim(0,255)
        #ax2.set_xticks(np.arange(0,256,15))
        ax2.grid('on')
        canvas2=FigureCanvasTkAgg(fig2,root)
        canvas2.get_tk_widget().place(x=950,y=285,width=500,height=200)
        canvas2.draw()   
        
        fig3=plt.figure() #ROI
        fig3.patch.set_facecolor('black')
        ax3=fig3.gca()
        ax3.axis('off')
        ax3.set_title("ROI",fontsize=10,color='white')
        canvas3=FigureCanvasTkAgg(fig3,root)
        canvas3.get_tk_widget().place(x=1075,y=15,width=250,height=250)
        canvas3.draw()   

        labROI=tk.Label(root,text="ROI")
        labROI["background"]="#9D9B9A"
        labROI["foreground"]="#000000"
        labROI["font"]=tkFont.Font(family='FreeSans',size=12)
        labROI["justify"]="center"
        labROI["relief"]="flat"
        labROI.place(x=800,y=520,width=100,height=20)
        
        labx=tk.Label(root,text="x:")
        #labx["background"]="#d3dbd6"
        labx["foreground"]="#000000"
        labx["font"]=tkFont.Font(family='FreeSans',size=10)
        labx["justify"]="center"
        labx["relief"]="flat"
        labx.place(x=800,y=550,width=50,height=20)
        
        px=tk.IntVar()
        px.set(100)
        entx = tk.Entry(root)
        entx["background"]="#ffffff"
        entx["foreground"]="#000000"
        entx["font"]=tkFont.Font(family='FreeSans',size=10)
        entx["borderwidth"]="1"
        entx["justify"]="center"
        entx["textvar"]=px
        entx["relief"]="sunken"
        entx.place(x=850,y=550,width=50,height=20)
        entx.bind('<Return>',self.ROIentry)
        #entx.bind('<FocusOut>',self.ROIentry)
        
        laby=tk.Label(root,text="y:")
        #laby["background"]="#d3dbd6"
        laby["foreground"]="#000000"
        laby["font"]=tkFont.Font(family='FreeSans',size=10)
        laby["justify"]="center"
        laby["relief"]="flat"
        laby.place(x=800,y=580,width=50,height=20)
        
        py=tk.IntVar()
        py.set(100)
        enty = tk.Entry(root)
        enty["background"]="#ffffff"
        enty["foreground"]="#000000"
        enty["font"]=tkFont.Font(family='FreeSans',size=10)
        enty["borderwidth"]="1"
        enty["justify"]="center"
        enty["textvar"]=py
        enty["relief"]="sunken"
        enty.place(x=850,y=580,width=50,height=20)
        enty.bind('<Return>',self.ROIentry)
        
        labw=tk.Label(root,text="ancho:")
        #labw["background"]="#d3dbd6"
        labw["foreground"]="#000000"
        labw["font"]=tkFont.Font(family='FreeSans',size=10)
        labw["justify"]="center"
        labw["relief"]="flat"
        labw.place(x=800,y=610,width=50,height=20)
        
        ancho=tk.IntVar()
        ancho.set(50)
        entw = tk.Entry(root)
        entw["background"]="#ffffff"
        entw["foreground"]="#000000"
        entw["font"]=tkFont.Font(family='FreeSans',size=10)
        entw["borderwidth"]="1"
        entw["justify"]="center"
        entw["textvar"]=ancho
        entw["relief"]="sunken"
        entw.place(x=850,y=610,width=50,height=20)
        entw.bind('<Return>',self.ROIentry)
        
        labh=tk.Label(root,text="alto:")
        #labh["background"]="#d3dbd6"
        labh["foreground"]="#000000"
        labh["font"]=tkFont.Font(family='FreeSans',size=10)
        labh["justify"]="center"
        labh["relief"]="flat"
        labh.place(x=800,y=640,width=50,height=20)
        
        alto=tk.IntVar()
        alto.set(50)
        enth = tk.Entry(root)
        enth["background"]="#ffffff"
        enth["foreground"]="#000000"
        enth["font"]=tkFont.Font(family='FreeSans',size=10)
        enth["borderwidth"]="1"
        enth["justify"]="center"
        enth["textvar"]=alto
        enth["relief"]="sunken"
        enth.place(x=850,y=640,width=50,height=20)
        enth.bind('<Return>',self.ROIentry)
        
        labmw=tk.Label(root,text="ancho:")
        #labh["background"]="#d3dbd6"
        labmw["foreground"]="#000000"
        labmw["font"]=tkFont.Font(family='FreeSans',size=10)
        labmw["justify"]="center"
        labmw["relief"]="flat"
        labmw.place(x=800,y=670,width=50,height=20)
        
        mw=tk.IntVar()
        mw.set(30)
        entmw = tk.Entry(root)
        entmw["background"]="#ffffff"
        entmw["foreground"]="#000000"
        entmw["font"]=tkFont.Font(family='FreeSans',size=10)
        entmw["borderwidth"]="1"
        entmw["justify"]="center"
        entmw["textvar"]=mw
        entmw["relief"]="sunken"
        entmw.place(x=850,y=670,width=50,height=20)
        entmw.bind('<Return>',self.ROIentry)
        
        labROIs=tk.Label(root) #ROI coordenadas
        #labROIs["background"]="#000000"
        #labROIs["foreground"]="#ffffff"
        labROIs["font"]=tkFont.Font(family='FreeSans',size=7)
        labROIs["anchor"]= "e"
        labROIs["relief"]="flat"
        labROIs.place(x=700,y=490,width=200,height=20)
        
        labaROI=tk.Label(root) #Numero de pixeles
        labaROI["background"]="#000000"
        labaROI["foreground"]="#ffffff"
        labaROI["font"]=tkFont.Font(family='FreeSans',size=7)
        labaROI["anchor"]= "w"
        labaROI["relief"]="flat"
        labaROI.place(x=1075,y=245,width=150,height=20)
        
        labherramientas=tk.Label(root,text="Presencia/ausencia")
        labherramientas["background"]="#9D9B9A"
        labherramientas["foreground"]="#000000"
        labherramientas["font"]=tkFont.Font(family='FreeSans',size=12)
        labherramientas["justify"]="left"
        labherramientas["relief"]="flat"
        labherramientas.place(x=50,y=545,width=160,height=25)
        
        Opciones=tk.IntVar()
        Opciones.set(1)  
        mbopcion=tk.Menubutton(root,text="Herramientas")
        mbopcionmenu=tk.Menu(mbopcion,tearoff=0)
        mbopcion["menu"]=mbopcionmenu
        mbopcionmenu.add_radiobutton(label="Presencia/ausencia", variable=Opciones,value=1,command=self.cambiaropciones)
        mbopcionmenu.add_radiobutton(label="Medición", variable=Opciones,value=2,command=self.cambiaropciones)
        mbopcion["background"]="#9D9B9A"
        mbopcion["foreground"]="#000000"
        mbopcion["activebackground"]="#9D9B9A"
        mbopcion["activeforeground"]="#000000"
        mbopcion["font"]=tkFont.Font(family='FreeSans',size=12)
        mbopcion["justify"]="left"
        mbopcion["relief"]="flat"
        mbopcion.place(x=50,y=520,width=160,height=25)      
        
        Herramienta=tk.IntVar()
        Herramienta.set(0)
        lbxherramienta=tk.Listbox(root,exportselection=False)
        lbxherramienta.insert(1,"Ninguna","Luminosidad","Recuento de píxeles","Blob","Circulo","Margen")
        lbxherramienta.selection_set(0)
        lbxherramienta["borderwidth"]=1
        lbxherramienta["foreground"]="#000000"
        lbxherramienta["font"]=tkFont.Font(family='FreeSans',size=10)
        lbxherramienta["justify"]="left"
        lbxherramienta["relief"]="raised"
        lbxherramienta["activestyle"]=tk.NONE
        lbxherramienta["selectbackground"]="#CAC8C8"
        lbxherramienta["selectforeground"]="#000000"
        lbxherramienta.place(x=50,y=580,width=150,height=150)
        lbxherramienta.bind('<<ListboxSelect>>',self.mostrartool)
        
        Medicion=tk.IntVar()
        Medicion.set(1)
        lbxMedicion=tk.Listbox(root,exportselection=False)
        lbxMedicion.insert(1,"Distancia circulo-circulo","Distancia circulo-línea","Distancia línea-línea","Angulo","Concentricidad")
        lbxMedicion.selection_set(0)
        lbxMedicion["borderwidth"]=1
        lbxMedicion["foreground"]="#000000"
        lbxMedicion["font"]=tkFont.Font(family='FreeSans',size=10)
        lbxMedicion["justify"]="left"
        lbxMedicion["relief"]="raised"
        lbxMedicion["activestyle"]=tk.NONE
        lbxMedicion["selectbackground"]="#CAC8C8"
        lbxMedicion["selectforeground"]="#000000"
        lbxMedicion.place(x=50,y=580,width=150,height=150)
        lbxMedicion.bind('<<ListboxSelect>>',self.mostrartool)
        
        scroll=tk.Scrollbar(root)
        scroll["orient"]=tk.VERTICAL
        lbxherramienta["yscrollcommand"]=scroll.set
        scroll.config(command=lbxherramienta.yview)
        scroll.place(x=200,y=580,width=10,height=150)
        
        labinspeccion=tk.Label(root,text="Inspección")
        labinspeccion["background"]="#9D9B9A"
        labinspeccion["foreground"]="#000000"
        labinspeccion["font"]=tkFont.Font(family='FreeSans',size=12)
        labinspeccion["justify"]="center"
        labinspeccion["relief"]="flat"
        labinspeccion.place(x=950,y=520,width=250,height=20)
        
        jeleccion=tk.IntVar()
        jeleccion.set(0)
        lbxinspeccion=tk.Listbox(root,exportselection=False)
        lbxinspeccion["borderwidth"]=1
        lbxinspeccion["foreground"]="#000000"
        lbxinspeccion["font"]=tkFont.Font(family='FreeSans',size=10)
        lbxinspeccion["justify"]="left"
        lbxinspeccion["relief"]="raised"
        lbxinspeccion["activestyle"]=tk.NONE
        lbxinspeccion["selectbackground"]="#CAC8C8"
        lbxinspeccion["selectforeground"]="#000000"
        lbxinspeccion.place(x=950,y=550,width=240,height=180)
        lbxinspeccion.bind('<<ListboxSelect>>',self.AbrirHerramienta)
        
        scroll2=tk.Scrollbar(root)
        scroll2["orient"]=tk.VERTICAL
        lbxinspeccion["yscrollcommand"]=scroll2.set
        scroll2.config(command=lbxinspeccion.yview)
        scroll2.place(x=1190,y=550,width=10,height=180)
       
        labmood=tk.Label(root,text="Modo")
        labmood["background"]="#9D9B9A"
        labmood["foreground"]="#000000"
        labmood["font"]=tkFont.Font(family='FreeSans',size=12)
        labmood["justify"]="left"
        labmood["relief"]="flat"
        labmood.place(x=50,y=15,width=160,height=20)
        
        Mood=tk.IntVar()
        Mood.set(1)
        rbm0=tk.Radiobutton(root,text="PC",variable=Mood,value=1,command=self.Modoadq)
        rbm0.place(x=50,y=45,width=160,height=20)
        rbm0["anchor"]= "w"
        rbm0["font"]=tkFont.Font(family='FreeSans',size=10)
        rbm1=tk.Radiobutton(root,text="Cámara",variable=Mood,value=2,command=self.Modoadq)
        rbm1.place(x=50,y=75,width=160,height=20)
        rbm1["anchor"]= "w"
        rbm1["font"]=tkFont.Font(family='FreeSans',size=10)
        
        but0= tk.Button(root,text="Cargar imágenes",command=self.Imagenes) 
        but0["activebackground"]="#FFFA77"
        but0["activeforeground"]="#000000"
        but0["background"]="#FCFB41"
        but0["foreground"]="#000000"
        but0["font"]=tkFont.Font(family='FreeSans',size=11)
        but0["justify"]="center"
        but0["relief"]="raised"
        but0.place(x=60,y=110,width=140,height=30)
        
        but4= tk.Button(root,text="Calibrar",command=self.Calibrar) 
        but4["activebackground"]="#FFFA77"
        but4["activeforeground"]="#000000"
        but4["background"]="#FCFB41"
        but4["foreground"]="#000000"
        but4["font"]=tkFont.Font(family='FreeSans',size=11)
        but4["justify"]="center"
        but4["relief"]="raised"
        but4.place(x=60,y=180,width=140,height=30)
                
        Gfoto=tk.BooleanVar()
        Gfoto.set(0)  
        rbgfoto=tk.Checkbutton(root,text="Guardar imagen",variable=Gfoto,command=self.GImagen)
        rbgfoto.place(x=50,y=150,width=160,height=20)
        rbgfoto["anchor"]= "w"
        rbgfoto["font"]=tkFont.Font(family='FreeSans',size=10)
                
        but2 = tk.Button(root,text="Previsualizar",command=self.ROI)
        but2["activebackground"]="#4C66FF"
        but2["activeforeground"]="#000000"
        but2["background"]="#3251CD"
        but2["foreground"]="#000000"
        but2["font"]=tkFont.Font(family='FreeSans',size=11)
        but2["justify"]="center"
        but2["relief"]="raised"
        but2.place(x=800,y=700,width=100,height=30)
        
        frame1=tk.Frame(root)
        #frame1["background"]="#000000"
        #frame1["relief"]="raised"
        frame1.place(x=250,y=510,width=495,height=220)
        
        but1= tk.Button(frame1,text="Analizar",command=self.Vision) 
        but1["activebackground"]="#FFBE4C"
        but1["activeforeground"]="#000000"
        but1["background"]="#F18905"
        but1["foreground"]="#000000"
        but1["font"]=tkFont.Font(family='FreeSans',size=11)
        but1["justify"]="center"
        but1["relief"]="raised"
        but1.place(x=365,y=150,width=100,height=30)
        
        but3= tk.Button(frame1,text="Guardar",command=self.GuardarHerramienta) 
        but3["activebackground"]="#FFBE4C"
        but3["activeforeground"]="#000000"
        but3["background"]="#F18905"
        but3["foreground"]="#000000"
        but3["font"]=tkFont.Font(family='FreeSans',size=11)
        but3["justify"]="center"
        but3["relief"]="raised"
        but3.place(x=365,y=190,width=100,height=30)

        labparametros=tk.Label(frame1,text="Parametros")
        labparametros["background"]="#9D9B9A"
        labparametros["foreground"]="#000000"
        labparametros["font"]=tkFont.Font(family='FreeSans',size=12)
        labparametros["justify"]="center"
        labparametros["relief"]="flat"
        labparametros.place(x=10,y=10,width=270,height=20)
        
        labtolmin=tk.Label(frame1,text="Tolerancia mínima:")
        #labx["background"]="#d3dbd6"
        labtolmin["foreground"]="#000000"
        labtolmin["font"]=tkFont.Font(family='FreeSans',size=10)
        labtolmin["anchor"]= "w"
        labtolmin["relief"]="flat"
        labtolmin.place(x=10,y=40,width=120,height=20)
        
        tolmin=tk.IntVar()
        tolmin.set(0)
        enttolmin = tk.Entry(frame1)
        enttolmin["background"]="#ffffff"
        enttolmin["foreground"]="#000000"
        enttolmin["font"]=tkFont.Font(family='FreeSans',size=10)
        enttolmin["borderwidth"]="1"
        enttolmin["justify"]="center"
        enttolmin["textvar"]=tolmin
        enttolmin["relief"]="sunken"
        enttolmin.place(x=130,y=40,width=50,height=20)
        enttolmin.bind('<Return>',self.Analizarentry)
        
        labtolmax=tk.Label(frame1,text="Tolerancia máxima:")
        #labx["background"]="#d3dbd6"
        labtolmax["foreground"]="#000000"
        labtolmax["font"]=tkFont.Font(family='FreeSans',size=10)
        labtolmax["anchor"]= "w"
        labtolmax["relief"]="flat"
        labtolmax.place(x=10,y=70,width=120,height=20)
        
        tolmax=tk.IntVar()
        tolmax.set(100)
        enttolmax = tk.Entry(frame1)
        enttolmax["background"]="#ffffff"
        enttolmax["foreground"]="#000000"
        enttolmax["font"]=tkFont.Font(family='FreeSans',size=10)
        enttolmax["borderwidth"]="1"
        enttolmax["justify"]="center"
        enttolmax["textvar"]=tolmax
        enttolmax["relief"]="sunken"
        enttolmax.place(x=130,y=70,width=50,height=20)
        enttolmax.bind('<Return>',self.Analizarentry)
        
        labpath=tk.Label(root,text=" ")
        #labpath["background"]="#000000"
        #labpath["foreground"]="#ffffff"
        labpath["font"]=tkFont.Font(family='FreeSans',size=7)
        labpath["anchor"]= "w"
        labpath["relief"]="flat"
        labpath.place(x=260,y=490,width=200,height=20)
        
        butmen= tk.Button(root,text="<",command=self.izquierda) 
        butmen["activebackground"]="#777777"
        butmen["activeforeground"]="#000000"
        butmen["background"]="#9D9B9A"
        butmen["foreground"]="#000000"
        butmen["font"]=tkFont.Font(family='FreeSans',size=11)
        butmen["justify"]="center"
        butmen["relief"]="raised"
        butmen.place(x=230,y=245,width=30,height=30)

        butmas = tk.Button(text=">",command=self.derecha)
        butmas["activebackground"]="#777777"
        butmas["activeforeground"]="#000000"
        butmas["background"]="#9D9B9A"
        butmas["foreground"]="#000000"
        butmas["font"]=tkFont.Font(family='FreeSans',size=11)
        butmas["justify"]="center"
        butmas["relief"]="raised"
        butmas.place(x=900,y=245,width=30,height=30)
        
        frame2=tk.Frame(root)
        #frame2["background"]="#000000"
        #frame1["relief"]="raised"
        frame2.place(x=250,y=600,width=290,height=120)
              
        labmodo=tk.Label(frame2,text="Modo:")
        #labx["background"]="#d3dbd6"
        labmodo["foreground"]="#000000"
        labmodo["font"]=tkFont.Font(family='FreeSans',size=10)
        labmodo["anchor"]= "w"
        labmodo["relief"]="flat"
        labmodo.place(x=10,y=10,width=40,height=20)
                
        modo=tk.IntVar()
        modo.set(1)
        rbmodo0=tk.Radiobutton(frame2,text="Automatico",variable=modo,value=1,command=self.mostrarumbral)
        rbmodo0.place(x=50,y=10,width=100,height=20)
        rbmodo0["anchor"]= "w"
        rbmodo0["font"]=tkFont.Font(family='FreeSans',size=10)
        rbmodo1=tk.Radiobutton(frame2,text="Manual",variable=modo,value=2,command=self.mostrarumbral)
        rbmodo1.place(x=150,y=10,width=100,height=20)
        rbmodo1["anchor"]= "w"
        rbmodo1["font"]=tkFont.Font(family='FreeSans',size=10)
        rbmodo=[rbmodo0,rbmodo1]
       
        labpolaridad=tk.Label(frame2,text="Polaridad:")
        #labx["background"]="#d3dbd6"
        labpolaridad["foreground"]="#000000"
        labpolaridad["font"]=tkFont.Font(family='FreeSans',size=10)
        labpolaridad["anchor"]= "w"
        labpolaridad["relief"]="flat"
        labpolaridad.place(x=10,y=70,width=70,height=20)
                
        polaridad=tk.IntVar()
        polaridad.set(1)
        rbpol0=tk.Radiobutton(frame2,text="Blanco",variable=polaridad,value=1,command=self.Vision)
        rbpol0.place(x=80,y=70,width=70,height=20)
        rbpol0["anchor"]= "w"
        rbpol0["font"]=tkFont.Font(family='FreeSans',size=10)
        rbpol1=tk.Radiobutton(frame2,text="Negro",variable=polaridad,value=2,command=self.Vision)
        rbpol1.place(x=150,y=70,width=70,height=20)
        rbpol1["anchor"]= "w"
        rbpol1["font"]=tkFont.Font(family='FreeSans',size=10)
        rbpol=[rbpol0,rbpol1]
        
        labumbral=tk.Label(frame2)
        #labumbral["background"]="#d3dbd6"
        labumbral["foreground"]="#000000"
        labumbral["font"]=tkFont.Font(family='FreeSans',size=10)
        labumbral["anchor"]= "w"
        labumbral["relief"]="flat"
        labumbral.place(x=10,y=40,width=80,height=20)

        umbral=tk.IntVar()
        umbral.set(127)
        entumb = tk.Entry(frame2)
        entumb["background"]="#ffffff"
        entumb["foreground"]="#000000"
        entumb["font"]=tkFont.Font(family='FreeSans',size=10)
        entumb["borderwidth"]="1"
        entumb["justify"]="center"
        entumb["textvar"]=umbral
        entumb["relief"]="sunken"
        entumb.place(x=90,y=40,width=50,height=20)
        entumb.bind('<Return>',self.Analizarentry)
                
        labumbalto=tk.Label(frame2,text="Umbral alto:")
        #labumbalto["background"]="#d3dbd6"
        labumbalto["foreground"]="#000000"
        labumbalto["font"]=tkFont.Font(family='FreeSans',size=10)
        labumbalto["anchor"]= "w"
        labumbalto["relief"]="flat"
        labumbalto.place(x=150,y=40,width=80,height=20)
        
        umbalto=tk.IntVar()
        umbalto.set(255)
        entumbalto = tk.Entry(frame2)
        entumbalto["background"]="#ffffff"
        entumbalto["foreground"]="#000000"
        entumbalto["font"]=tkFont.Font(family='FreeSans',size=10)
        entumbalto["borderwidth"]="1"
        entumbalto["justify"]="center"
        entumbalto["textvar"]=umbalto
        entumbalto["relief"]="sunken"
        entumbalto.place(x=230,y=40,width=50,height=20)
        entumbalto.bind('<Return>',self.Analizarentry)
        
        labumbral.config(state="disabled")
        entumb.config(state="disabled")
        labumbalto.config(state="disabled")
        entumbalto.config(state="disabled")
        
        labRes=tk.Label(frame1,text="Resultados")
        labRes["background"]="#9D9B9A"
        labRes["foreground"]="#000000"
        labRes["font"]=tkFont.Font(family='FreeSans',size=12)
        labRes["justify"]="center"
        labRes["relief"]="flat"
        labRes.place(x=345,y=10,width=140,height=20)
        
        labvalor=tk.Label(frame1)
        #labx["background"]="#d3dbd6"
        labvalor["foreground"]="#000000"
        labvalor["font"]=tkFont.Font(family='FreeSans',size=10)
        labvalor["anchor"]= "w"
        labvalor["relief"]="flat"
        labvalor.place(x=345,y=40,width=90,height=20)
        
        valor=tk.StringVar()
        valor.set('_____')
        labrvalor = tk.Label(frame1)
        labrvalor["background"]="#ffffff"
        labrvalor["foreground"]="#000000"
        labrvalor["font"]=tkFont.Font(family='FreeSans',size=10)
        labrvalor["justify"]="right"
        labrvalor["textvariable"]=valor
        labrvalor["relief"]="flat"
        labrvalor.place(x=435,y=40,width=50,height=20)
                
        labresult=tk.Label(frame1,text="Pieza ")
        #labx["background"]="#d3dbd6"
        labresult["foreground"]="#000000"
        labresult["font"]=tkFont.Font(family='FreeSans',size=10)
        labresult["anchor"]= "w"
        labresult["relief"]="flat"
        labresult.place(x=345,y=70,width=50,height=20)
        
        Resultado=tk.StringVar()
        Resultado.set('_______')
        labrresult = tk.Label(frame1)
        labrresult["background"]="#ffffff"
        labrresult["foreground"]="#000000"
        labrresult["font"]=tkFont.Font(family='FreeSans',size=10)
        labrresult["justify"]="center"
        labrresult["textvariable"]=Resultado
        labrresult["relief"]="flat"
        labrresult.place(x=395,y=70,width=90,height=20)
        
        Resultado2=tk.StringVar()
        Resultado2.set('_____________')
        labrresult2 = tk.Label(frame1)
        labrresult2["background"]="#ffffff"
        labrresult2["foreground"]="#000000"
        labrresult2["font"]=tkFont.Font(family='FreeSans',size=10)
        labrresult2["justify"]="center"
        labrresult2["textvariable"]=Resultado2
        labrresult2["relief"]="flat"
        labrresult2.place(x=345,y=100,width=140,height=40)
        
        frame3=tk.Frame(root)
        #frame3["background"]="#000000"
        #frame3["relief"]="raised"
        frame3.place(x=250,y=600,width=290,height=120)
              
        labmodmed=tk.Label(frame3,text="Seleccionar:")
        #labmodmed["background"]="#d3dbd6"
        labmodmed["foreground"]="#000000"
        labmodmed["font"]=tkFont.Font(family='FreeSans',size=10)
        labmodmed["anchor"]= "w"
        labmodmed["relief"]="flat"
        labmodmed.place(x=10,y=10,width=80,height=20)
                
        Figura=tk.IntVar()
        Figura.set(1)
        rbfig0=tk.Radiobutton(frame3,text="Circulo 1",variable=Figura,value=1,command=self.ROIFigura)
        rbfig0.place(x=100,y=10,width=100,height=20)
        rbfig0["anchor"]= "w"
        rbfig0["font"]=tkFont.Font(family='FreeSans',size=10)
        rbfig1=tk.Radiobutton(frame3,text="Circulo 2",variable=Figura,value=2,command=self.ROIFigura)
        rbfig1.place(x=200,y=10,width=100,height=20)
        rbfig1["anchor"]= "w"
        rbfig1["font"]=tkFont.Font(family='FreeSans',size=10)
        rbfig=[rbfig0,rbfig1]
        
                
        labtmed=tk.Label(frame3,text="Tipo de medición:")
        #labx["background"]="#d3dbd6"
        labtmed["foreground"]="#000000"
        labtmed["font"]=tkFont.Font(family='FreeSans',size=10)
        labtmed["anchor"]= "w"
        labtmed["relief"]="flat"
        labtmed.place(x=10,y=40,width=120,height=20)
                
        tmedicion=tk.IntVar()
        tmedicion.set(1)  
        rbtmed0=tk.Radiobutton(frame3,text="Punto medio a línea",variable=tmedicion,value=1,command=self.Vision)
        rbtmed0.place(x=10,y=70,width=230,height=20)
        rbtmed0["anchor"]= "w"
        rbtmed0["font"]=tkFont.Font(family='FreeSans',size=10)
        rbtmed1=tk.Radiobutton(frame3,text="Punto medio a punto medio",variable=tmedicion,value=2,command=self.Vision)
        rbtmed1.place(x=10,y=100,width=230,height=20)
        rbtmed1["anchor"]= "w"
        rbtmed1["font"]=tkFont.Font(family='FreeSans',size=10)
        rbtmed=[rbtmed0,rbtmed1]
        
        
        ROI1=[200,200,75,75,30]
        ROI2=[100,100,50,50,30]
        
        frame1.place_forget()
        frame2.place_forget()
        frame3.place_forget() 
        labmw.place_forget()   
        entmw.place_forget() 
        lbxMedicion.place_forget()  
        butmen.place_forget()
        butmas.place_forget()
        rbgfoto.place_forget()

        menubar=tk.Menu(root)
        filemenu=tk.Menu(menubar,tearoff=0)
        filemenu.add_command(label="Nuevo",command=self.create_widgets)
        filemenu.add_command(label="Guardar inspección",command=self.Guardar)
        filemenu.add_command(label="Abrir inspección",command=self.Abrir)
        filemenu.add_command(label="Guardar calibración",command=self.Guardarcal)
        filemenu.add_command(label="Abrir calibración",command=self.Abrircal)
        menubar.add_cascade(label="Archivo", menu=filemenu)
        root.config(menu=menubar)
    
    def Guardar(self):
        global insp,directorio,Mood
        modo=Mood.get() 
        if (insp.shape[0]>0):
            if (modo==1):
                nuevafila=pd.DataFrame([[directorio]],columns=['Carpeta'])
            insp=pd.concat([insp,nuevafila], ignore_index=True)
            directorioinsp = filedialog.asksaveasfilename(defaultextension=".xlsx",filetypes=[("Excel Workbook", "*.xlsx")])
            insp.to_excel(directorioinsp, index=False, sheet_name='Hoja1')
        else:
            tk.messagebox.showwarning(message="Configura una inspección antes de guardar")
        
    def Abrir(self):
        global insp,directorio,abin,j
        self.create_widgets()
        directorioinsp = filedialog.askopenfilename(filetypes=[("Excel Workbook", "*.xlsx")])
        insp=pd.read_excel(directorioinsp, sheet_name='Hoja1', header=0,decimal=',')
        #print(insp)
        
        #Revisar
        filas=insp.shape[0]
        directorio=insp.loc[filas-1,'Carpeta']
        #print(directorio)
        abin=1
        self.Imagenes()
        
        filas=np.arange(0,insp.shape[0]-1,1)
        for a in filas:
            tool=insp.loc[a,'Herramienta']
            if (tool==1): #Luminosidad
                nombre='Luminosidad'
            elif (tool==2): #Recuento de pixeles
                nombre='Recuento de píxeles'
            elif (tool==3): #Blob
                nombre="Blob"
            elif (tool==4): #Circulo
                nombre="Circulo"
            elif (tool==5): #Margen/Linea
                nombre="Margen"
            elif (tool==6):
                nombre="Distancia circulo-circulo"
            elif (tool==7): 
                nombre="Distancia circulo-línea"
            elif (tool==8): 
                nombre="Distancia línea-línea"
            elif (tool==9): 
                nombre="Angulo"
            elif (tool==10): 
                nombre="Concentricidad"
            self.PanelInspeccion(nombre,a)
        lbxinspeccion.selection_clear(0, tk.END)
        lbxinspeccion.selection_set(0)
        j=insp.shape[0]-1
        insp=insp.drop(insp.shape[0]-1)
        self.Abrirlasherramientas()   
        
    def Guardarcal(self):
        global Ls
        if (len(Ls)!=0):
            LsDf=pd.DataFrame([[Ls[0],Ls[1],Ls[2],Ls[3],Ls[4],Ls[5],Ls[6],Ls[7],Ls[8],Ls[9],Ls[10]]],columns=['L1','L2','L3','L4','L5','L6','L7','L8','L9','L10','L11'])
            directoriocal = filedialog.asksaveasfilename(defaultextension=".xlsx",filetypes=[("Excel Workbook", "*.xlsx")])
            LsDf.to_excel(directoriocal, index=False, sheet_name='Hoja1')
        else:
            tk.messagebox.showwarning(message="Configura una calibración antes de guardar")
        
    def Abrircal(self):
        global Ls,Herramienta,Opciones
        choose=Opciones.get() 
        tool=Herramienta.get()
        directoriocal = filedialog.askopenfilename(filetypes=[("Excel Workbook", "*.xlsx")])
        LsDf=pd.read_excel(directoriocal, sheet_name='Hoja1', header=0,decimal=',')
        Ls=LsDf.to_numpy()[0]
        #print(Ls)
        if (tool!=0 or choose==2):
            self.Vision()   
        
    def PanelInspeccion(self,t,n):
        texto=str(n+1)+'. '+t    
        lbxinspeccion.insert(n,texto)
        
    def Modoadq(self):
        global Mood,but0,butmen,butmas,labpath,vid,rbgfoto,Gfoto
        modo=Mood.get() 
        img=np.zeros([480,640])
        self.MostrarImg(img)
        butmen.place_forget()
        butmas.place_forget()
        labpath.place_forget()
        if (modo==1): 
            but0.config(text="Cargar imágenes")
            vid.release()
            rbgfoto.place_forget()
        elif (modo==2):    
            but0.config(text="Disparador")  
            #vid=cv2.VideoCapture(0) #0 webcam, 1 camara
            Gfoto.set(0)
            rbgfoto.place(x=50,y=150,width=160,height=20)

            #rbm1.config(state="disabled")
        
    def cambiaropciones(self):
        global Opciones,labherramientas,lbxherramienta,lbxMedicion
        choose=Opciones.get() 
        if (choose==1):
            lbxherramienta.place(x=50,y=580,width=150,height=150)
            lbxMedicion.place_forget()   
            labherramientas.config(text="Presencia/ausencia")
        elif (choose==2):    
            lbxherramienta.place_forget()
            lbxMedicion.place(x=50,y=580,width=150,height=150) 
            labherramientas.config(text="Medición")
        self.mostrar()
        
    def mostrartool(self,event):
        global Opciones,lbxherramienta,Herramienta,lbxMedicion,Medicion
        choose=Opciones.get() 
        if (choose==1):
            if (lbxherramienta.curselection()):
                opcion=lbxherramienta.curselection()[0]
                Herramienta.set(opcion)
            #texto=lbxherramienta.get(opcion)
            #print(texto)
        elif (choose==2):  
            if (lbxMedicion.curselection()):
                opcion=lbxMedicion.curselection()[0]
                Medicion.set(opcion+1)
            #texto=lbxMedicion.get(opcion)
            #print(texto)
        self.mostrar()
        
    def mostrar(self):
        global Opciones,Herramienta,Medicion,Figura,rbfig,rbtmed,labtmed,frame1,frame2,frame3,labvalor,labtolmin,labtolmax,enttolmax,labumbral,labumbalto,entumbalto,labx,laby,labw,labh,labmw,entmw,labpolaridad,rbpol,labrresult2
        choose=Opciones.get() 
        tool=Herramienta.get() 
        if (choose==1):  
            frame3.place_forget()
            if (tool==0): 
                frame1.place_forget()
                frame2.place_forget()
            elif(tool==1):
                labvalor.config(text="Luminosidad:")
                frame1.place(x=250,y=510,width=495,height=220)
                frame2.place_forget()
            elif(tool==2):
                labvalor.config(text="Pixeles:")
                labumbral.config(text="Umbral bajo:")
                frame1.place(x=250,y=510,width=495,height=220)
                frame2.place(x=250,y=600,width=290,height=120)
                labumbalto.place(x=150,y=40,width=80,height=20)
                entumbalto.place(x=230,y=40,width=50,height=20)
            elif(tool==3):
                labvalor.config(text="Resultado:")
                labumbral.config(text="Umbral:")
                frame1.place(x=250,y=510,width=495,height=220)
                frame2.place(x=250,y=600,width=290,height=120)
                labumbalto.place_forget()
                entumbalto.place_forget()
            elif(tool==4): 
                labvalor.config(text="Resultado:")
                labumbral.config(text="Umbral:")
                frame1.place(x=250,y=510,width=495,height=220)
                frame2.place(x=250,y=600,width=290,height=120)
                labumbalto.place_forget()
                entumbalto.place_forget()
                labw.config(text="r max:")
                labh.config(text="r min:")
                labpolaridad.place_forget()
                rbpol[0].place_forget()
                rbpol[1].place_forget()
                labrresult2.place(x=345,y=100,width=140,height=40) 
            elif(tool==5): 
                labvalor.config(text="Resultado:")
                labumbral.config(text="Umbral bajo:")
                frame1.place(x=250,y=510,width=495,height=220)
                frame2.place(x=250,y=600,width=290,height=120)
                labumbalto.place(x=150,y=40,width=80,height=20)
                entumbalto.place(x=230,y=40,width=50,height=20)
                labx.config(text="x1:")
                laby.config(text="y1:")
                labw.config(text="x2:")
                labh.config(text="y2:")
                labpolaridad.place_forget()
                rbpol[0].place_forget()
                rbpol[1].place_forget()
                labrresult2.place(x=345,y=100,width=140,height=40) 
                labmw.place(x=800,y=670,width=50,height=20)
                entmw.place(x=850,y=670,width=50,height=20)
                labtolmax.place_forget()
                enttolmax.place_forget()
                labtolmin.config(text="Rango del ángulo:")
            if(tool<4):
                labw.config(text="ancho:")
                labh.config(text="alto:")
                labpolaridad.place(x=10,y=70,width=70,height=20)
                rbpol[0].place(x=80,y=70,width=70,height=20)
                rbpol[1].place(x=150,y=70,width=70,height=20)
                labrresult2.place_forget()   
            if(tool!=5):   
                labx.config(text="x:")
                laby.config(text="y:")
                labmw.place_forget()   
                entmw.place_forget()   
                labtolmax.place(x=10,y=70,width=120,height=20)
                enttolmax.place(x=130,y=70,width=50,height=20)
                labtolmin.config(text="Tolerancia mínima:")
        if (choose==2):
            med=Medicion.get()  
            figure=Figura.get()
            frame2.place_forget()
            frame3.place(x=250,y=600,width=290,height=120)  
            labvalor.config(text="Distancia(pix):")
            frame1.place(x=250,y=510,width=495,height=220)
            labrresult2.place(x=345,y=100,width=140,height=40) 
            if(tool==5):
                labtolmax.place(x=10,y=70,width=120,height=20)
                enttolmax.place(x=130,y=70,width=50,height=20)
                labtolmin.config(text="Tolerancia mínima:")
            #if (med==0): 
                #frame1.place_forget() 
                #frame3.place_forget() 
            if (med==4):
                labvalor.config(text="Ángulo(grados):")
                frame3.place(x=250,y=600,width=290,height=120)  
            if(med==1 or med==5):
                labx.config(text="x:")
                laby.config(text="y:")
                labw.config(text="r max:")
                labh.config(text="r min:")
                labmw.place_forget()   
                entmw.place_forget()  
                rbfig[0].config(text="Circulo 1")
                rbfig[1].config(text="Circulo 2")
            elif(med==2):
                rbfig[0].config(text="Circulo")
                rbfig[1].config(text="Línea")  
                if (figure==1):
                    labx.config(text="x:")
                    laby.config(text="y:")
                    labw.config(text="r max:")
                    labh.config(text="r min:")
                    labmw.place_forget()   
                    entmw.place_forget()  
                elif (figure==2):
                    labx.config(text="x1:")
                    laby.config(text="y1:")
                    labw.config(text="x2:")
                    labh.config(text="y2:")
                    labmw.place(x=800,y=670,width=50,height=20)
                    entmw.place(x=850,y=670,width=50,height=20)
            elif(med==3 or med==4):
                labx.config(text="x1:")
                laby.config(text="y1:")
                labw.config(text="x2:")
                labh.config(text="y2:")
                labmw.place(x=800,y=670,width=50,height=20)
                entmw.place(x=850,y=670,width=50,height=20)
                rbfig[0].config(text="Línea 1")
                rbfig[1].config(text="Línea 2") 
            if (med==2):
                rbtmed[0].place(x=10,y=70,width=230,height=20)
                rbtmed[1].place(x=10,y=100,width=230,height=20)
                labtmed.place(x=10,y=40,width=120,height=20)
                rbtmed[0].config(text="Centro del circulo a línea")
                rbtmed[1].config(text="Centro a punto medio de la línea") 
            elif (med==3):
                rbtmed[0].place(x=10,y=70,width=230,height=20)
                rbtmed[1].place(x=10,y=100,width=230,height=20)
                labtmed.place(x=10,y=40,width=120,height=20)
                rbtmed[0].config(text="Punto medio a línea")
                rbtmed[1].config(text="Punto medio a punto medio") 
            elif (med==4):
                labrresult2.place_forget()  
            else:
                rbtmed[0].place_forget() 
                rbtmed[1].place_forget() 
                labtmed.place_forget()  
            
        self.ROI()
        self.Tolerancias()
        valor.set("____")
        Resultado.set("_____")
        self.Vision()
    
    def Tolerancias(self):
        global Opciones,Herramienta,Medicion,tolmin,tolmax,ancho,alto
        choose=Opciones.get() 
        w=ancho.get()
        h=alto.get()
        a=w*h
        tool=Herramienta.get()   
        if (choose==1):
            if(tool==1):
                tolmin.set(15)
                tolmax.set(240) 
            elif(tool==2):
                tolmin.set(round(a*0.1))
                tolmax.set(round(a*0.9))
            elif(tool==3):
                tolmin.set(round(a*0.1))
                tolmax.set(round(a*0.9))
            elif(tool==4):
                tolmin.set(h)
                tolmax.set(w)
            elif(tool==5):
                tolmin.set(10)
        elif (choose==2): #Revisar
            med=Medicion.get()
            if (med==5):
                tolmin.set(0)
                tolmax.set(5) 
            elif (med==4):
                tolmin.set(45)
                tolmax.set(135) 
            else:
                tolmin.set(10)
                tolmax.set(500) 
            
    def mostrarumbral(self):
        global modo,labumbral,entumb,labumbalto,entumbalto
        mode=modo.get()
        if (mode==1):
            labumbral.config(state="disabled")
            entumb.config(state="disabled")
            labumbalto.config(state="disable")
            entumbalto.config(state="disable")  
        elif(mode==2):
            labumbral.config(state="normal")
            entumb.config(state="normal")         
            labumbalto.config(state="normal")
            entumbalto.config(state="normal")
        self.Vision()
    
    def ROIFigura(self):
        global Figura,ROI1,ROI2,px,py,ancho,alto,mw,Medicion
        med=Medicion.get()    
        figure=Figura.get()
        #print(figure)
        if (figure==1):
            ROI2=[px.get(),py.get(),ancho.get(),alto.get(),mw.get()]
            px.set(ROI1[0])
            py.set(ROI1[1])
            ancho.set(ROI1[2])
            alto.set(ROI1[3])
            mw.set(ROI1[4])
        elif (figure==2):
            ROI1=[px.get(),py.get(),ancho.get(),alto.get(),mw.get()]
            px.set(ROI2[0])
            py.set(ROI2[1])
            ancho.set(ROI2[2])
            alto.set(ROI2[3])
            mw.set(ROI2[4])
            
        if(med==1 or med==5):
            labx.config(text="x:")
            laby.config(text="y:")
            labw.config(text="r max:")
            labh.config(text="r min:")
            labmw.place_forget()   
            entmw.place_forget() 
        elif(med==2):
            if (figure==1):
                labx.config(text="x:")
                laby.config(text="y:")
                labw.config(text="r max:")
                labh.config(text="r min:")
                labmw.place_forget()   
                entmw.place_forget()  
            elif (figure==2):
                labx.config(text="x1:")
                laby.config(text="y1:")
                labw.config(text="x2:")
                labh.config(text="y2:")
                labmw.place(x=800,y=670,width=50,height=20)
                entmw.place(x=850,y=670,width=50,height=20)
        elif(med==3 or med==4):
            labx.config(text="x1:")
            laby.config(text="y1:")
            labw.config(text="x2:")
            labh.config(text="y2:")
            labmw.place(x=800,y=670,width=50,height=20)
            entmw.place(x=850,y=670,width=50,height=20)
        self.Vision()
        
    def Calibrar(self):
        global pixeles,centimetros,entry_matriz,top,img
        corners = cv2.goodFeaturesToTrack(img, 7, 0.01, 5)
        corners = np.int0(corners)
        #print(corners)
        pixeles=[]
        m=1
        imgRGB=cv2.cvtColor(img,cv2.COLOR_GRAY2RGB)
        for i in corners:
            x,y = i.ravel()
            cv2.circle(imgRGB, (x,y), 3, (0,0,255), -1)
            cv2.putText(imgRGB,str(m),(x,y),2,1,(0,0,255),1)
            pixeles.append([x,y]) 
            m=m+1
        cv2.line(imgRGB, (0,0),(0,img.shape[0]), (0, 255, 0), 6, cv2.LINE_AA)
        cv2.putText(imgRGB,'Eje V',(10,img.shape[0]-10),1,1,(0,255,0),1)
        cv2.line(imgRGB, (0,0),(img.shape[1],0), (255, 0, 0), 6, cv2.LINE_AA)
        cv2.putText(imgRGB,'Eje U',(img.shape[1]-50,20),1,1,(255,0,0),1)
        cv2.circle(imgRGB, (0,0), 5, (0,0,255), -1)
        cv2.putText(imgRGB,'(0,0)pixeles',(10,20),1,1,(0,0,255),1)
        self.MostrarImg(imgRGB)

        top = tk.Toplevel(root)
        top.title("Calibración")

        guardar_button = tk.Button(top, text="Guardar", command=self.guardar_datos)
        guardar_button.grid(row=2, column=0)
        
        # Crea una tabla en un widget Frame
        frame = tk.Frame(top)
        frame.grid(row=0, column=0)

        # Crea las etiquetas para las columnas de la tabla
        tk.Label(frame, text="Punto").grid(row=1, column=0)
        tk.Label(frame, text="Coordenadas pixeles").grid(row=0, column=1,columnspan=2)
        tk.Label(frame, text="Coordenadas centimetros").grid(row=0, column=3,columnspan=2)
        tk.Label(frame, text="u").grid(row=1, column=1)
        tk.Label(frame, text="v").grid(row=1, column=2)
        tk.Label(frame, text="x").grid(row=1, column=3)
        tk.Label(frame, text="y").grid(row=1, column=4)
        # Crea las etiquetas para las filas de la tabla
        tk.Label(frame, text="1").grid(row=2, column=0)
        tk.Label(frame, text="2").grid(row=3, column=0)
        tk.Label(frame, text="3").grid(row=4, column=0)
        tk.Label(frame, text="4").grid(row=5, column=0)
        tk.Label(frame, text="5").grid(row=6, column=0)
        tk.Label(frame, text="6").grid(row=7, column=0)
        tk.Label(frame, text="7").grid(row=8, column=0)

        # Crea la tabla con los datos de la matriz en dos columnas
        for i, fila in enumerate(pixeles):
            tk.Label(frame, text=fila[0],width=10).grid(row=i+2, column=1)
            tk.Label(frame, text=fila[1],width=10).grid(row=i+2, column=2)

        # Crea los campos Entry para ingresar los datos de la nueva matriz
        entry_matriz = []
        for i in range(7):
            fila_entry = []
            for j in range(2):
                entry = tk.Entry(frame,width=10)
                entry.grid(padx=2,row=i+2, column=j+3)
                fila_entry.append(entry)
            entry_matriz.append(fila_entry)

        top.mainloop()
        
    def Imagenes(self):
        global abin,insp,contenido,i,n,labpath,img,directorio,Mood,vid,butmen,labpath,butmas,Herramienta,Opciones
        choose=Opciones.get() 
        tool=Herramienta.get()
        modo=Mood.get() 
        if (modo==1):         
            i=0
            #if (insp.shape==0):
            if (abin==0):
                directorio = filedialog.askdirectory()
            elif (abin==1):
                abin=0
            contenido = os.listdir(directorio)
            for filename in os.listdir(directorio):
                # Convertir el nombre de archivo a UTF-8
                filename = filename.encode('utf-8').decode('utf-8')
            
            
            path=directorio+'/'+contenido[0]
            img_array = np.fromfile(path, np.uint8) #Opción para leer imagenes con directorios con acentos o ñ
            imgBGR = cv2.imdecode(img_array, cv2.IMREAD_UNCHANGED)
            #imgBGR=cv2.imread(path,1)
            imgRGB=cv2.cvtColor(imgBGR,cv2.COLOR_BGR2RGB)
            img=cv2.cvtColor(imgBGR,cv2.COLOR_BGR2GRAY)
            self.MostrarImg(img)

            labpath.config(text=f"Imagen 1: {contenido[0]}")
            labpath.place(x=260,y=490,width=200,height=20)

            if (len(contenido)>1):
                butmen.place(x=230,y=245,width=30,height=30)
                butmas.place(x=900,y=245,width=30,height=30)
        elif (modo==2):   
            guardar=Gfoto.get() 
            vid=cv2.VideoCapture(0) #0 webcam, 1 camara
            #vid=cv2.VideoCapture("http://192.168.174.84:8080/videofeed")
            ret,imgBGR=vid.read()
            vid.release()
            imgRGB=cv2.cvtColor(imgBGR,cv2.COLOR_BGR2RGB)
            img=cv2.cvtColor(imgBGR,cv2.COLOR_BGR2GRAY)
            self.MostrarImg(img)
            if(guardar==1):
                hora_actual = datetime.datetime.now()
                momento = hora_actual.strftime('%d%m%Y_%H%M%S')
                nombre='Camara'+momento+'.jpg'
                labpath.config(text=f"Imagen {n}: {nombre}")
                path=directorio+'/'+nombre
                #path=filedialog.asksaveasfilename(defaultextension=".jpg",filetypes=[("jpeg files", "*.jpg")])
                cv2.imwrite(path,img)
                contenido = os.listdir(directorio)
                i=n-1
                n=n+1
        if (tool!=0 or choose==2):
            self.Vision()
            
            
    def GImagen(self):
        global directorio,img,i,n,Gfoto,contenido
        guardar=Gfoto.get() 
        if(guardar==1):
            directorio = filedialog.askdirectory()
            contenido = os.listdir(directorio)
            i=len(contenido)
            n=len(contenido)+1
            butmen.place(x=230,y=245,width=30,height=30)
            butmas.place(x=900,y=245,width=30,height=30)
            labpath.place(x=260,y=490,width=200,height=20)
            labpath.config(text="")
        else:
            #directorio=' '
            butmen.place_forget()
            butmas.place_forget()
            labpath.place_forget()

    def MostrarImg(self,img):
        global im2,canvas
        #Imagen centrada (hacer compensación de coordenadas)
        im1=Image.fromarray(img)
        im2= ImageTk.PhotoImage(image=im1)
        canvas.create_image(640/2,480/2,image=im2)  
            
    def izquierda(self):
        global directorio,contenido,i,labpath,img,Herramienta,Opciones
        choose=Opciones.get() 
        tool=Herramienta.get()
        i=i-1
        if (i<0):
            i=len(contenido)-1
        path=directorio+'/'+contenido[i]
        labpath.config(text=f"Imagen {i+1}: {contenido[i]}")
        img_array = np.fromfile(path, np.uint8)
        imgBGR = cv2.imdecode(img_array, cv2.IMREAD_UNCHANGED)
        #imgBGR=cv2.imread(path,1)
        imgRGB=cv2.cvtColor(imgBGR,cv2.COLOR_BGR2RGB)
        img=cv2.cvtColor(imgBGR,cv2.COLOR_BGR2GRAY)
        self.MostrarImg(img)
        if (tool!=0 or choose==2):
            self.Vision()
        #else:
            #self.ROI()
            
    def derecha(self):
        global directorio,contenido,i,labpath,img,Herramienta,Opciones
        choose=Opciones.get() 
        tool=Herramienta.get()
        i=i+1
        if (i>=len(contenido)):
            i=0
        path=directorio+'/'+contenido[i]
        labpath.config(text=f"Imagen {i+1}: {contenido[i]}")
        img_array = np.fromfile(path, np.uint8)
        imgBGR = cv2.imdecode(img_array, cv2.IMREAD_UNCHANGED)
        #imgBGR=cv2.imread(path,1)
        imgRGB=cv2.cvtColor(imgBGR,cv2.COLOR_BGR2RGB)
        img=cv2.cvtColor(imgBGR,cv2.COLOR_BGR2GRAY)
        self.MostrarImg(img)
        if (tool!=0 or choose==2):
            self.Vision()
        #else:
            #self.ROI()
            
    def AbrirHerramienta(self,event):
        self.Abrirlasherramientas()
        
    def Abrirlasherramientas(self):
        global insp,Opciones,Herramienta,Medicion,ROI1,ROI2,mbopcionmenu,lbxinspeccion,lbxherramienta,lbxMedicion,tmedicion,img,px,py,ancho,alto,mw,tolmin,tolmax,modo,polaridad,umbral,umbalto,labROIs,labaROI,valor,Resultado,rbmodo,rbpol 
        #print(insp)
        opcion=lbxinspeccion.curselection()[0]
        tool=insp.loc[opcion,'Herramienta']
        #print(tool)
        if (tool<6):
            Opciones.set(1) 
            mbopcionmenu.invoke(0)
            Herramienta.set(tool)
            lbxherramienta.selection_clear(0, tk.END) 
        elif (tool>=6):
            Opciones.set(2)
            mbopcionmenu.invoke(1)
            Figura.set(1)
            Medicion.set(tool-5)
            lbxMedicion.selection_clear(0, tk.END)
        if (tool==1): #Luminosidad
            lbxherramienta.selection_set(1)
        elif (tool==2): #Recuento de pixeles
            lbxherramienta.selection_set(2)
        elif (tool==3): #Blob
            lbxherramienta.selection_set(3)
        elif (tool==4): #Circulo
            lbxherramienta.selection_set(4)
        elif (tool==5): #Margen
            lbxherramienta.selection_set(5) 
        elif (tool==6): #Distancia circulo-circulo
                lbxMedicion.selection_set(0)
        elif (tool==7): #"Distancia circulo-linea
                lbxMedicion.selection_set(1)
        elif (tool==8): #Distancia linea-linea
                lbxMedicion.selection_set(2)
        elif (tool==9): #Angulo
                lbxMedicion.selection_set(3)
        elif (tool==10): #Concentricidad
                lbxMedicion.selection_set(4) 
        ROI1=[100,100,50,50,30]
        ROI2=[100,100,50,50,30]
        
        x=int(insp.loc[opcion,'x'])
        px.set(x)
        y=int(insp.loc[opcion,'y'])
        py.set(y)
        w=int(insp.loc[opcion,'ancho'])
        ancho.set(w)
        h=int(insp.loc[opcion,'alto'])
        alto.set(h)
        self.cambiaropciones()
        
        tol_min=insp.loc[opcion,'Tolerancia minima']
        tolmin.set(tol_min)
        if (tool!=5):
            tol_max=insp.loc[opcion,'Tolerancia maxima']
            tolmax.set(tol_max)
        if (tool==2): 
            ppolaridad=insp.loc[opcion,'Polaridad']
            polaridad.set(ppolaridad)
            rbpol[int(ppolaridad-1)].select()
            pmodo=insp.loc[opcion,'Modo']
            modo.set(pmodo)
            rbmodo[int(pmodo-1)].select()
            umb=insp.loc[opcion,'Umbral']
            umbral.set(umb)
            umb_alto=insp.loc[opcion,'Umbral alto']
            umbalto.set(umb_alto)
        elif (tool==3):
            ppolaridad=insp.loc[opcion,'Polaridad']
            polaridad.set(ppolaridad)
            rbpol[int(ppolaridad-1)].select()
            pmodo=insp.loc[opcion,'Modo']
            modo.set(pmodo)
            rbmodo[int(pmodo-1)].select()
            umb=insp.loc[opcion,'Umbral']
            umbral.set(umb)           
        elif (tool==4):
            pmodo=insp.loc[opcion,'Modo']
            modo.set(pmodo)
            rbmodo[int(pmodo-1)].select()
            umb=insp.loc[opcion,'Umbral']
            umbral.set(umb)
        elif (tool==5):
            pmodo=insp.loc[opcion,'Modo']
            modo.set(pmodo)
            rbmodo[int(pmodo-1)].select()
            umb=insp.loc[opcion,'Umbral']
            umbral.set(umb)
            umb_alto=insp.loc[opcion,'Umbral alto']
            umbalto.set(umb_alto)
            mancho=int(insp.loc[opcion,'grosor'])
            mw.set(mancho)        
        elif (tool>=6):
            tol_max=insp.loc[opcion,'Tolerancia maxima']
            tolmax.set(tol_max)
            mancho=int(insp.loc[opcion,'grosor'])
            mw.set(mancho) 
            tipol=int(insp.loc[opcion,'Tipo'])
            tmedicion.set(tipol) 
            ROI1=[x,y,w,h,mancho]
            x1=int(insp.loc[opcion,'x1'])
            y1=int(insp.loc[opcion,'y1'])
            x2=int(insp.loc[opcion,'x2'])
            y2=int(insp.loc[opcion,'y2'])
            grosor=int(insp.loc[opcion,'grosor2'])
            ROI2=[x1,y1,x2,y2,grosor]
        #valor.set("____")
        #Resultado.set("_____")
        self.mostrarumbral()
        self.Vision()
            

    def GuardarHerramienta(self):
        global Opciones,j,insp,lbxherramienta,lbxMedicion,tmedicion,lbxinspeccion,Herramienta,Medicion,ROI1,ROI2,img,px,py,ancho,alto,mw,tolmin,tolmax,modo,polaridad,umbral,umbalto,labROIs,labaROI,valor,Resultado
        choose=Opciones.get() 
        if (j==0):
            insp=pd.DataFrame(columns=['Herramienta','x','y','ancho','alto','grosor','Tolerancia minima','Tolerancia maxima','Polaridad','Modo','Umbral','Umbral alto','x1','y1','x2','y2','grosor2','Tipo','Carpeta'])
        if (choose==1):
            tool=Herramienta.get()
            x=px.get()
            y=py.get()
            w=ancho.get()
            h=alto.get()
            mancho=mw.get()
            #ROI=[x,y,w,h,mancho]
            tol_min=tolmin.get()
            tol_max=tolmax.get()
            pmodo=modo.get()
            ppolaridad=polaridad.get()
            umb=umbral.get()
            umb_alto=umbalto.get()
            if (tool==1): #Luminosidad
                nuevafila=pd.DataFrame([[tool,x,y,w,h,tol_min,tol_max]],columns=['Herramienta','x','y','ancho','alto','Tolerancia minima','Tolerancia maxima'])
            elif (tool==2): #Recuento de pixeles
                nuevafila=pd.DataFrame([[tool,x,y,w,h,tol_min,tol_max,ppolaridad,pmodo,umb,umb_alto]],columns=['Herramienta','x','y','ancho','alto','Tolerancia minima','Tolerancia maxima','Polaridad','Modo','Umbral','Umbral alto'])
            elif (tool==3): #Blob
                nuevafila=pd.DataFrame([[tool,x,y,w,h,tol_min,tol_max,ppolaridad,pmodo,umb]],columns=['Herramienta','x','y','ancho','alto','Tolerancia minima','Tolerancia maxima','Polaridad','Modo','Umbral'])
            elif (tool==4): #Circulo
                nuevafila=pd.DataFrame([[tool,x,y,w,h,tol_min,tol_max,pmodo,umb]],columns=['Herramienta','x','y','ancho','alto','Tolerancia minima','Tolerancia maxima','Modo','Umbral'])
            elif (tool==5): #Margen
                nuevafila=pd.DataFrame([[tool,x,y,w,h,mancho,tol_min,pmodo,umb,umb_alto]],columns=['Herramienta','x','y','ancho','alto','grosor','Tolerancia minima','Modo','Umbral','Umbral alto'])
            insp=pd.concat([insp,nuevafila], ignore_index=True)
            #print(insp)
            opcion=lbxherramienta.curselection()[0]
            nombre=lbxherramienta.get(opcion)  
        if (choose==2):
            med=Medicion.get()
            figure=Figura.get()  
            tol_min=tolmin.get()
            tol_max=tolmax.get()
            tipol=tmedicion.get()
            if (figure==1):
                ROI1=[px.get(),py.get(),ancho.get(),alto.get(),mw.get()]
            elif (figure==2):
                ROI2=[px.get(),py.get(),ancho.get(),alto.get(),mw.get()]       
            nuevafila=pd.DataFrame([[med+5,ROI1[0],ROI1[1],ROI1[2],ROI1[3],ROI1[4],tol_min,tol_max,ROI2[0],ROI2[1],ROI2[2],ROI2[3],ROI2[4],tipol]],columns=['Herramienta','x','y','ancho','alto','grosor','Tolerancia minima','Tolerancia maxima','x1','y1','x2','y2','grosor2','Tipo'])
            insp=pd.concat([insp,nuevafila], ignore_index=True)
            opcion=lbxMedicion.curselection()[0]
            nombre=lbxMedicion.get(opcion)  
        self.PanelInspeccion(nombre,j)
        lbxinspeccion.selection_clear(0, tk.END)
        lbxinspeccion.selection_set(j)
        j=j+1
                
    def Analizarentry(self,event):
        self.Vision()
    
    def Vision(self):
        global Ls,calibracion,labrresult2,Opcion,Herramienta,Medicion,Figura,tmedicion,ROI1,ROI2,img,px,py,ancho,mw,alto,tolmin,tolmax,modo,polaridad,umbral,umbalto,labROIs,labaROI,valor,Resultado,Resultado2
        choose=Opciones.get() 
        tol_min=tolmin.get()
        tol_max=tolmax.get() 
        if(choose==1):
            tool=Herramienta.get()
            x=px.get()
            y=py.get()
            w=ancho.get()
            h=alto.get()
            mancho=mw.get()
            ROI=[x,y,w,h,mancho]
            tol_min=tolmin.get()
            tol_max=tolmax.get()
            pmodo=modo.get()
            ppolaridad=polaridad.get()
            umb=umbral.get()
            umb_alto=umbalto.get()
            labROIs.config(text=f"ROI=( {x}, {y}, {w}, {h})")
            labaROI.config(text=f"{w*h} píxeles")

            if ppolaridad==1:
                ppolaridad='blanco'
            elif ppolaridad==2:
                ppolaridad='negro'

            if pmodo==1:
                pmodo='auto'
            elif pmodo==2:
                pmodo='manual' 

            Coor=' '
            if (tool==1):
                Val,Res=self.Luminosidad(img,ROI,tol_min,tol_max)
            elif (tool==2):
                Val,Res=self.RecPixeles(img,ROI,tol_min,tol_max,ppolaridad,pmodo,umb,umb_alto)
            elif (tool==3):
                Res=self.Blob(img,ROI,tol_min,tol_max,ppolaridad,pmodo,umb)
                if (Res==1):
                    Val="P" #Presente
                elif(Res==0):
                    Val="NP" #No presente
            elif (tool==4):
                labROIs.config(text=f"ROI=c:({x}, {y}) rmin:{h} rmax:{h}")
                Res,xcirculo,ycirculo,radio=self.Circulo(img,ROI,tol_min,tol_max,pmodo,umb)
                if (Res==1):
                    Val="P" #Presente
                    Coor='c:('+str(xcirculo)+', '+str(ycirculo)+'), r:'+str(radio)
                    if(len(Ls)!= 0):
                        u1=int(xcirculo)
                        v1=int(ycirculo)
                        u2=int(xcirculo+radio)
                        v2=int(ycirculo)
                        A=np.array([[(u1*Ls[8])-Ls[0],(u1*Ls[9])-Ls[1],(u1*Ls[10])-Ls[2]],
                                    [(v1*Ls[8])-Ls[4],(v1*Ls[9])-Ls[5],(v1*Ls[10])-Ls[6]]])
                        B=np.array([[Ls[3]-u1],
                                    [Ls[7]-v1]])
                        Api=np.linalg.pinv(A)
                        cm1=Api@B
                        C=np.array([[(u2*Ls[8])-Ls[0],(u2*Ls[9])-Ls[1],(u2*Ls[10])-Ls[2]],
                                        [(v2*Ls[8])-Ls[4],(v2*Ls[9])-Ls[5],(v2*Ls[10])-Ls[6]]])
                        D=np.array([[Ls[3]-u2],
                                        [Ls[7]-v2]])
                        Cpi=np.linalg.pinv(C)
                        cm2=Cpi@D
                        rcm=math.sqrt((cm1[0]-cm2[0])**2+(cm1[1]-cm2[1])**2)
                        rcm=round(rcm,2)
                        xcm=round(float(cm1[0]),2)
                        ycm=round(float(cm1[1]),2)
                        labrresult2.config(font=tkFont.Font(size=8))
                        Coor='pix: c:('+str(xcirculo)+', '+str(ycirculo)+'), r:'+str(radio)+'\ncm: c:('+str(xcm)+', '+str(ycm)+'), r:'+str(rcm)
                    else:
                        labrresult2.config(font=tkFont.Font(size=10))
                elif(Res==0):
                        Val="NP" #No presente 
            elif (tool==5):
                labROIs.config(text=f"ROI=P1:({x}, {y}) P2:({w}, {h}) w:{mancho}")
                Res,x1linea,y1linea,x2linea,y2linea=self.Linea(img,ROI,tol_min,pmodo,umb,umb_alto)
                if (Res==1):
                    Val="P" #Presente
                    Coor='P1:('+str(x1linea)+', '+str(y1linea)+')\nP2:('+str(x2linea)+', '+str(y2linea)+')'
                    if(len(Ls)!= 0):
                        u1=int(x1linea)
                        v1=int(y1linea)
                        u2=int(x2linea)
                        v2=int(y2linea)
                        A=np.array([[(u1*Ls[8])-Ls[0],(u1*Ls[9])-Ls[1],(u1*Ls[10])-Ls[2]],
                                    [(v1*Ls[8])-Ls[4],(v1*Ls[9])-Ls[5],(v1*Ls[10])-Ls[6]]])
                        B=np.array([[Ls[3]-u1],
                                    [Ls[7]-v1]])
                        Api=np.linalg.pinv(A)
                        cm1=Api@B
                        C=np.array([[(u2*Ls[8])-Ls[0],(u2*Ls[9])-Ls[1],(u2*Ls[10])-Ls[2]],
                                        [(v2*Ls[8])-Ls[4],(v2*Ls[9])-Ls[5],(v2*Ls[10])-Ls[6]]])
                        D=np.array([[Ls[3]-u2],
                                        [Ls[7]-v2]])
                        Cpi=np.linalg.pinv(C)
                        cm2=Cpi@D
                        labrresult2.config(font=tkFont.Font(size=6))
                        Coor='pix: P1:('+str(x1linea)+', '+str(y1linea)+') P2:('+str(x2linea)+', '+str(y2linea)+')'+'\ncm: P1:('+str(round(float(cm1[0]),2))+', '+str(round(float(cm1[1]),2))+') P2:('+str(round(float(cm2[0]),2))+', '+str(round(float(cm2[1]),2))+')'
                    else:
                        labrresult2.config(font=tkFont.Font(size=10))
                elif(Res==0):
                    Val="NP" #No presente
            if (Res==1):
                pRes="Aceptada"
            elif(Res==0):
                pRes="Rechazada"
            valor.set(Val)
            Resultado.set(pRes)
            Resultado2.set(Coor)
            
        if(choose==2):
            imgRGB=cv2.cvtColor(img,cv2.COLOR_GRAY2RGB)
            med=Medicion.get() 
            figure=Figura.get()
            tipo=tmedicion.get()
            Resultado2.set('')
            labrresult2.config(font=tkFont.Font(size=10))
            if (figure==1):
                ROI1=[px.get(),py.get(),ancho.get(),alto.get(),mw.get()]
            elif (figure==2):
                ROI2=[px.get(),py.get(),ancho.get(),alto.get(),mw.get()]
            labaROI.config(text=f"{ancho.get()*alto.get()} píxeles")
            ROI=[ROI1,ROI2]
            ax3.set_title("ROI Bordes Canny",fontsize=10,color='white')
            distancia=[]        
            if(med==1 or med==5):
                labROIs.config(text=f"ROI=c:({px.get()}, {py.get()}) rmin:{alto.get()} rmax:{ancho.get()}")
                Res1,x1,y1,r1,imgR1,imgROI1,hist1=self.DCirculo(imgRGB,ROI1)
                Res2,x2,y2,r2,imgR2,imgROI2,hist2=self.DCirculo(imgR1,ROI2)
                if (Res1==1 and Res2==1):
                    distancia=math.sqrt((int(x1)-int(x2))**2+(int(y1)-int(y2))**2)
                    distancia=round(distancia,2)
                    if(len(Ls)!= 0):
                        u1=int(x1)
                        v1=int(y1)
                        u2=int(x2)
                        v2=int(y2)
                        A=np.array([[(u1*Ls[8])-Ls[0],(u1*Ls[9])-Ls[1],(u1*Ls[10])-Ls[2]],
                                    [(v1*Ls[8])-Ls[4],(v1*Ls[9])-Ls[5],(v1*Ls[10])-Ls[6]]])
                        B=np.array([[Ls[3]-u1],
                                    [Ls[7]-v1]])
                        Api=np.linalg.pinv(A)
                        cm1=Api@B
                        C=np.array([[(u2*Ls[8])-Ls[0],(u2*Ls[9])-Ls[1],(u2*Ls[10])-Ls[2]],
                                    [(v2*Ls[8])-Ls[4],(v2*Ls[9])-Ls[5],(v2*Ls[10])-Ls[6]]])
                        D=np.array([[Ls[3]-u2],
                                    [Ls[7]-v2]])
                        Cpi=np.linalg.pinv(C)
                        cm2=Cpi@D
                        distanciacm=math.sqrt((cm1[0]-cm2[0])**2+(cm1[1]-cm2[1])**2)
                        #print(distanciacm)
                        distanciacm=round(float(distanciacm),2)
                        Resultado2.set('Distancia(cm): '+str(distanciacm)) 
                    if (distancia>tol_min and distancia<tol_max):
                        cv2.line(imgR2, (x1,y1),(x2,y2), (0, 255, 0), 2, cv2.LINE_AA)
                        pRes="Aceptada"
                        #imgR=self.ImgRes(imgR2,ROI,'green')
                    else:
                        cv2.line(imgR2, (x1,y1),(x2,y2), (255, 0, 0), 2, cv2.LINE_AA)
                        #imgR=self.ImgRes(imgR2,ROI,'red')
                        pRes="Rechazada"
                else:
                    pRes="Rechazada"
            elif(med==2):
                if (figure==1):
                    labROIs.config(text=f"ROI=c:({px.get()}, {py.get()}) rmin:{alto.get()} rmax:{ancho.get()}")
                elif (figure==2):
                    labROIs.config(text=f"ROI=P1:({px.get()}, {py.get()}) P2:({ancho.get()}, {alto.get()}) w:{mw.get()}")
                Res1,xc,yc,r,imgR1,imgROI1,hist1=self.DCirculo(imgRGB,ROI1)
                Res2,x1,y1,x2,y2,imgR2,imgROI2,hist2=self.DLinea(imgR1,ROI2)
                if (Res1==1 and Res2==1):
                    if (tipo==1):#Distancia centro a linea
                        #Linea y=m(x-x0)+y0 --> y=mx-mx0+y0 --> mx-y-mx0+y0=0
                        m=(y2-y1)/(x2-x1) 
                        #Distancia de a recta Ax+By+C=0 al punto (x0,y0)
                        #Ax+By+C=0--> mx-y-mx0+y0=0 --> A=m B=-1 C=-mx0+y0=0
                        #d=|(Ax0+By0+C)/sqrt(A**2+B**2)|
                        distancia=abs(((m*xc)+(-1*yc)+(-(m*x1)+y1))/(math.sqrt((m)**2+(-1)**2)))
                        distancia=round(distancia,2)
                        #Linea perpendicular a l2 y=m(x-x0)+y0 --> y=mx-mx0+y0 --> -mx+y=-mx0+y0
                        mp=-1/m
                        #Ax+By=C--> -mx+y=-mx0+y0 --> A=-m B=1 C=-mx0+y0=0
                        A=np.array([[-mp,1],[-m,1]])
                        B=np.array([-(mp*xc)+yc,-(m*x1)+y1])
                        coor=np.linalg.solve(A, B)
                        if(len(Ls)!= 0):
                            u1=int(xc)
                            v1=int(yc)
                            u2=int(x1)
                            v2=int(y1)
                            A=np.array([[(u1*Ls[8])-Ls[0],(u1*Ls[9])-Ls[1],(u1*Ls[10])-Ls[2]],
                                        [(v1*Ls[8])-Ls[4],(v1*Ls[9])-Ls[5],(v1*Ls[10])-Ls[6]]])
                            B=np.array([[Ls[3]-u1],
                                        [Ls[7]-v1]])
                            Api=np.linalg.pinv(A)
                            cm1=Api@B
                            C=np.array([[(u2*Ls[8])-Ls[0],(u2*Ls[9])-Ls[1],(u2*Ls[10])-Ls[2]],
                                        [(v2*Ls[8])-Ls[4],(v2*Ls[9])-Ls[5],(v2*Ls[10])-Ls[6]]])
                            D=np.array([[Ls[3]-u2],
                                        [Ls[7]-v2]])
                            Cpi=np.linalg.pinv(C)
                            cm2=Cpi@D
                            distanciacm=abs(((m*cm1[0])+(-1*cm1[1])+(-(m*cm2[0])+cm2[1]))/(math.sqrt((m)**2+(-1)**2)))
                            distanciacm=round(float(distanciacm),2)
                            Resultado2.set('Distancia(cm): '+str(distanciacm)) 
                        if (distancia>tol_min and distancia<tol_max):
                            cv2.line(imgR2, (xc,yc),(int(coor[0]),int(coor[1])), (0, 255, 0), 2, cv2.LINE_AA)
                            pRes="Aceptada"
                        else:
                            cv2.line(imgR2, (xc,yc),(int(coor[0]),int(coor[1])), (255, 0, 0), 2, cv2.LINE_AA)
                            pRes="Rechazada"
                    elif (tipo==2): #Distancia centro a punto medio
                        #Punto medio
                        xm=int((x1+x2)/2)
                        ym=int((y1+y2)/2)
                        distancia=math.sqrt((int(xc)-int(xm))**2+(int(yc)-int(ym))**2)
                        distancia=round(distancia,2)
                        if(len(Ls)!= 0):
                            u1=int(xc)
                            v1=int(yc)
                            u2=int(xm)
                            v2=int(ym)
                            A=np.array([[(u1*Ls[8])-Ls[0],(u1*Ls[9])-Ls[1],(u1*Ls[10])-Ls[2]],
                                        [(v1*Ls[8])-Ls[4],(v1*Ls[9])-Ls[5],(v1*Ls[10])-Ls[6]]])
                            B=np.array([[Ls[3]-u1],
                                        [Ls[7]-v1]])
                            Api=np.linalg.pinv(A)
                            cm1=Api@B
                            C=np.array([[(u2*Ls[8])-Ls[0],(u2*Ls[9])-Ls[1],(u2*Ls[10])-Ls[2]],
                                        [(v2*Ls[8])-Ls[4],(v2*Ls[9])-Ls[5],(v2*Ls[10])-Ls[6]]])
                            D=np.array([[Ls[3]-u2],
                                        [Ls[7]-v2]])
                            Cpi=np.linalg.pinv(C)
                            cm2=Cpi@D
                            distanciacm=math.sqrt((cm1[0]-cm2[0])**2+(cm1[1]-cm2[1])**2)
                            distanciacm=round(float(distanciacm),2)
                            Resultado2.set('Distancia(cm): '+str(distanciacm)) 
                        if (distancia>tol_min and distancia<tol_max):
                            cv2.line(imgR2, (xc,yc),(xm,ym), (0, 255, 0), 2, cv2.LINE_AA)
                            pRes="Aceptada"
                        else:
                            cv2.line(imgR2, (xc,yc),(xm,ym), (255, 0, 0), 2, cv2.LINE_AA)
                            pRes="Rechazada"
                else:
                    pRes="Rechazada"
            elif(med==3):
                labROIs.config(text=f"ROI=P1:({px.get()}, {py.get()}) P2:({ancho.get()}, {alto.get()}) w:{mw.get()}")
                Res1,x11,y11,x21,y21,imgR1,imgROI1,hist1=self.DLinea(imgRGB,ROI1)
                Res2,x12,y12,x22,y22,imgR2,imgROI2,hist2=self.DLinea(imgR1,ROI2)
                if (Res1==1 and Res2==1):
                    if (tipo==1): #Distancia punto medio a linea
                        #Punto medio
                        xm=int((x11+x21)/2)
                        ym=int((y11+y21)/2)
                        #Linea y=m(x-x0)+y0 --> y=mx-mx0+y0 --> mx-y-mx0+y0=0
                        m=(y22-y12)/(x22-x12) 
                        #Distancia de a recta Ax+By+C=0 al punto (x0,y0)
                        #Ax+By+C=0--> mx-y-mx0+y0=0 --> A=m B=-1 C=-mx0+y0=0
                        #d=|(Ax0+By0+C)/sqrt(A**2+B**2)|
                        distancia=abs(((m*xm)+(-1*ym)+(-(m*x12)+y12))/(math.sqrt((m)**2+(-1)**2)))
                        distancia=round(distancia,2)
                        #Linea perpendicular a l2 y=m(x-x0)+y0 --> y=mx-mx0+y0 --> -mx+y=-mx0+y0
                        mp=-1/m
                        #Ax+By=C--> -mx+y=-mx0+y0 --> A=-m B=1 C=-mx0+y0=0
                        A=np.array([[-mp,1],[-m,1]])
                        B=np.array([-(mp*xm)+ym,-(m*x12)+y12])
                        coor=np.linalg.solve(A, B)
                        if(len(Ls)!= 0):
                            u1=int(xm)
                            v1=int(ym)
                            u2=int(x12)
                            v2=int(y12)
                            A=np.array([[(u1*Ls[8])-Ls[0],(u1*Ls[9])-Ls[1],(u1*Ls[10])-Ls[2]],
                                        [(v1*Ls[8])-Ls[4],(v1*Ls[9])-Ls[5],(v1*Ls[10])-Ls[6]]])
                            B=np.array([[Ls[3]-u1],
                                        [Ls[7]-v1]])
                            Api=np.linalg.pinv(A)
                            cm1=Api@B
                            C=np.array([[(u2*Ls[8])-Ls[0],(u2*Ls[9])-Ls[1],(u2*Ls[10])-Ls[2]],
                                        [(v2*Ls[8])-Ls[4],(v2*Ls[9])-Ls[5],(v2*Ls[10])-Ls[6]]])
                            D=np.array([[Ls[3]-u2],
                                        [Ls[7]-v2]])
                            Cpi=np.linalg.pinv(C)
                            cm2=Cpi@D
                            distanciacm=abs(((m*cm1[0])+(-1*cm1[1])+(-(m*cm2[0])+cm2[1]))/(math.sqrt((m)**2+(-1)**2)))
                            distanciacm=round(float(distanciacm),2)
                            Resultado2.set('Distancia(cm): '+str(distanciacm)) 
                        if (distancia>tol_min and distancia<tol_max):
                            cv2.line(imgR2, (xm,ym),(int(coor[0]),int(coor[1])), (0, 255, 0), 2, cv2.LINE_AA)
                            pRes="Aceptada"
                        else:
                            cv2.line(imgR2, (xm,ym),(int(coor[0]),int(coor[1])), (255, 0, 0), 2, cv2.LINE_AA)
                            pRes="Rechazada"
                    elif (tipo==2): #Distancia punto medio a punto medio
                        #Puntos medios
                        xm1=int((x11+x21)/2)
                        ym1=int((y11+y21)/2)
                        xm2=int((x12+x22)/2)
                        ym2=int((y12+y22)/2)
                        distancia=math.sqrt((int(xm1)-int(xm2))**2+(int(ym1)-int(ym2))**2)
                        distancia=round(distancia,2)
                        if(len(Ls)!= 0):
                            u1=int(xm1)
                            v1=int(ym1)
                            u2=int(xm2)
                            v2=int(ym2)
                            A=np.array([[(u1*Ls[8])-Ls[0],(u1*Ls[9])-Ls[1],(u1*Ls[10])-Ls[2]],
                                        [(v1*Ls[8])-Ls[4],(v1*Ls[9])-Ls[5],(v1*Ls[10])-Ls[6]]])
                            B=np.array([[Ls[3]-u1],
                                        [Ls[7]-v1]])
                            Api=np.linalg.pinv(A)
                            cm1=Api@B
                            C=np.array([[(u2*Ls[8])-Ls[0],(u2*Ls[9])-Ls[1],(u2*Ls[10])-Ls[2]],
                                        [(v2*Ls[8])-Ls[4],(v2*Ls[9])-Ls[5],(v2*Ls[10])-Ls[6]]])
                            D=np.array([[Ls[3]-u2],
                                        [Ls[7]-v2]])
                            Cpi=np.linalg.pinv(C)
                            cm2=Cpi@D
                            distanciacm=math.sqrt((cm1[0]-cm2[0])**2+(cm1[1]-cm2[1])**2)
                            distanciacm=round(float(distanciacm),2)
                            Resultado2.set('Distancia(cm): '+str(distanciacm)) 
                        if (distancia>tol_min and distancia<tol_max):
                            cv2.line(imgR2, (xm1,ym1),(xm2,ym2), (0, 255, 0), 2, cv2.LINE_AA)
                            pRes="Aceptada"
                        else:
                            cv2.line(imgR2, (xm1,ym1),(xm2,ym2), (255, 0, 0), 2, cv2.LINE_AA)
                            pRes="Rechazada"
                else:
                    pRes="Rechazada"
            elif (med==4):
                labROIs.config(text=f"ROI=P1:({px.get()}, {py.get()}) P2:({ancho.get()}, {alto.get()}) w:{mw.get()}")
                Res1,x11,y11,x21,y21,imgR1,imgROI1,hist1=self.DLinea(imgRGB,ROI1)
                Res2,x12,y12,x22,y22,imgR2,imgROI2,hist2=self.DLinea(imgR1,ROI2)
                if (Res1==1 and Res2==1):
                    #Angulo entre 2 rectas theta=atan((m2-m1)/(1+m1m2))
                    #Pendientes
                    m1=(y21-y11)/(x21-x11) 
                    m2=(y22-y12)/(x22-x12)     
                    if (m1==m2):
                        distancia=0
                    elif (m1*m2==-1):
                        distancia=90
                    else:
                        distancia=abs(math.degrees(math.atan((m2-m1)/(1+m1*m2))))
                    distancia=round(distancia,2)
                    #Puntos medios
                    xm1=int((x11+x21)/2)
                    ym1=int((y11+y21)/2)
                    xm2=int((x12+x22)/2)
                    ym2=int((y12+y22)/2)
                    if (distancia>tol_min and distancia<tol_max):
                        cv2.line(imgR2, (xm1,ym1),(xm2,ym2), (0, 255, 0), 2, cv2.LINE_AA)
                        pRes="Aceptada"
                    else:
                        cv2.line(imgR2, (xm1,ym1),(xm2,ym2), (255, 0, 0), 2, cv2.LINE_AA)
                        pRes="Rechazada"
                else:
                    pRes="Rechazada"
            valor.set(distancia)
            Resultado.set(pRes)

            self.MostrarImg(imgR2)
            ax2.clear()
            if (figure==1):
                self.MostrarROI(imgROI1)
                ordhist=sorted(hist1, reverse=True)
                maxhist=ordhist[1]
                ax2.plot(hist1,'k')      
            elif (figure==2):
                self.MostrarROI(imgROI2)
                ordhist=sorted(hist2, reverse=True)
                maxhist=ordhist[1]
                ax2.plot(hist2,'k')
            ax2.set_title("Histograma de la ROI",fontsize=10)
            ax2.set_xlabel("Nivel de gris")
            ax2.set_ylabel("Pixeles")
            ax2.set_ylim([0,maxhist+10])
            ax2.set_xlim(0,255)
            ax2.grid('on')
            canvas2.draw()    
            
    def Luminosidad(self,img,ROI,tolmin,tolmax):
        x=ROI[0]
        y=ROI[1]
        w=ROI[2]
        h=ROI[3]
        #Extraer la ROI
        imgROI=img[y:y+h,x:x+w]
        #Crear el histograma de la ROI
        hist=cv2.calcHist([imgROI],[0],None,[256],[0,256])
        #Calcular la luminosidad de la ROI
        g=np.arange(0,256,1)
        #g=np.linspace(0,255,256)
        M=w*h
        #M=imgROI.size
        Lum=(g.dot(hist))/M
        #Determinar si la pieza es aceptada o no según las tolerancias
        if Lum>tolmin:
            if Lum<tolmax:
                Res=1 #Aceptado
            else:
                Res=0 #Rechazado
        else:
            Res=0 #Rechazado 
            
        #Mostrar la imagen con el resultado
        if Res==1:
            imgR=self.ImgRes(img,ROI,'green')
        else:
            imgR=self.ImgRes(img,ROI,'red')
        self.MostrarImg(imgR)
        Lum=float(Lum)

        #Graficar histograma
        ax2.clear()
        ax2.plot(hist,'k')
        ax2.axvline(x=tolmin,color='r')
        ax2.text(tolmin-10,max(hist)+10, tolmin,fontsize=10,color='r')
        ax2.axvline(x=tolmax,color='r')
        ax2.text(tolmax-10,max(hist)+10, tolmax,fontsize=10,color='r')
        ax2.text(Lum-10,max(hist)+10, round(Lum,2),fontsize=10,color='b')
        ax2.axvline(x=Lum,color='b')
        ax2.set_title("Histograma de la ROI",fontsize=10)
        ax2.set_xlabel("Nivel de gris")
        ax2.set_ylabel("Pixeles")
        ax2.set_xlim(0,255)
        ax2.grid('on')
        canvas2.draw()
        
        #ROI más grande
        self.MostrarROI(imgROI)
        return round(Lum,2),Res

    def RecPixeles(self,img,ROI,tolmin=0,tolmax=100,polaridad='blanco',modo='auto',ubajo=127,ualto=255):
        x=ROI[0]
        y=ROI[1]
        w=ROI[2]
        h=ROI[3]
        #Extraer la ROI
        imgROI=img[y:y+h,x:x+w]
        #Crear el histograma de la ROI
        hist=cv2.calcHist([imgROI],[0],None,[256],[0,256])
        if modo=='auto':
            #Umbralizar por Otsu
            ubajo,imgbn=cv2.threshold(imgROI,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
        elif modo=='manual':
            #Umbralizar por Rango Umbral
            imgbn=cv2.inRange(imgROI,ubajo,ualto)
        #Numero de pixeles blancos
        pblanco=(sum(sum(imgbn,0.0)))/255
        if polaridad=='blanco':
            pixeles=pblanco
        elif polaridad=='negro':
            #Pixeles totales
            #M=w*h
            M=imgbn.size
            #Numero de pixeles negros
            pixeles=M-pblanco      
        #Determinar si la pieza es aceptada o no según las tolerancias
        if pixeles>tolmin:
            if pixeles<tolmax:
                Res=1 #Aceptado
            else:
                Res=0 #Rechazado
        else:
            Res=0 #Rechazado 
        #Remplazar ROI de imagen original con imagen umbralizada
        imgEdit=img.copy()
        imgEdit[y:y+h,x:x+w]=imgbn
        
        #Mostrar la imagen con el resultado
        if Res==1:
            imgR=self.ImgRes(imgEdit,ROI,'green')
        else:
            imgR=self.ImgRes(imgEdit,ROI,'red')
        self.MostrarImg(imgR)
        
        #Graficar histograma
        ax2.clear()
        ax2.plot(hist,'k')
        ax2.axvline(x=ubajo,color='b')
        ax2.text(ubajo-10,max(hist)+10, ubajo,fontsize=10,color='b')
        if modo=='manual':
            ax2.axvline(x=ualto,color='b')
            ax2.text(ualto-10,max(hist)+10, ualto,fontsize=10,color='b')
        ax2.set_title("Histograma de la ROI",fontsize=10)
        ax2.set_xlabel("Nivel de gris")
        ax2.set_ylabel("Pixeles")
        ax2.set_xlim(0,255)
        ax2.grid('on')
        canvas2.draw()
        
        #ROI más grande
        self.MostrarROI(imgbn)
        ax3.set_title("ROI umbralizado",fontsize=10,color='white')
        return int(pixeles),Res
    
    def Blob(self,img,ROI,tolmin=0,tolmax=100,polaridad='blanco',modo='auto',umb=127):
        x=ROI[0]
        y=ROI[1]
        w=ROI[2]
        h=ROI[3]
        #Extraer la ROI
        imgROI=img[y:y+h,x:x+w]
        #Crear el histograma de la ROI
        hist=cv2.calcHist([imgROI],[0],None,[256],[0,256])
        if modo=='auto':
            #Umbralizar por Otsu
            umb,imgbn=cv2.threshold(imgROI,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
        elif modo=='manual':
            #Umbralizar por Umbral Manual
            th,imgbn=cv2.threshold(imgROI,umb,255,cv2.THRESH_BINARY)
        #Blob
        if polaridad=='blanco':
               contours,hierarchy=cv2.findContours(imgbn,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
        elif polaridad=='negro': #Sacar negativo de la imagen (pues cv2.findContours detecta objetos blancos)
                if modo=='auto':
                    umb,imgbninv=cv2.threshold(imgROI,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
                elif modo=='manual':
                    th,imgbninv=cv2.threshold(imgROI,umb,255,cv2.THRESH_BINARY_INV)
                contours,hierarchy=cv2.findContours(imgbninv,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
        contours = sorted(contours, key=cv2.contourArea, reverse=True) #Ordena por area, reverse=True se utiliza para ordenar los contornos en orden descendente
        imgEdit=img.copy()
        #Dibujar contornos con areas dentro de las tolerancias en azul
        Res=0
        imgR=self.ImgRes(imgEdit,ROI,'red')
        #ROI más grande
        self.MostrarROI(imgbn)
        ax3.set_title("ROI umbralizado",fontsize=10,color='white')
        for i in range (len(contours)):
            area = cv2.contourArea(contours[i])
            if (tolmin<=area<=tolmax):
                imgR=self.ImgRes(imgEdit,ROI,'green')
                cv2.drawContours(imgR[y:y+h,x:x+w,:],contours,i,(0,0,255),2)
                imgblobROI=cv2.cvtColor(imgbn,cv2.COLOR_GRAY2RGB)
                cv2.drawContours(imgblobROI,contours,i,(0,0,255),1)
                Res=1
                #ROI más grande
                self.MostrarROI(imgblobROI)
                ax3.set_title("ROI umbralizado",fontsize=10,color='white')
                break #para que muestre el area más grande
        
        #Mostrar la imagen con el resultado
        self.MostrarImg(imgR)
        
        #Graficar histograma
        ax2.clear()
        ax2.plot(hist,'k')
        ax2.axvline(x=umb,color='b')
        ax2.text(umb-10,max(hist)+10, umb,fontsize=10,color='b')
        ax2.set_title("Histograma de la ROI",fontsize=10)
        ax2.set_xlabel("Nivel de gris")
        ax2.set_ylabel("Pixeles")
        ax2.set_xlim(0,255)
        ax2.grid('on')
        canvas2.draw()
        return Res
            
    def Circulo(self,img,ROI,tolmin=0,tolmax=0,modo='auto',umb=127):
        x=ROI[0]
        y=ROI[1]
        r=ROI[2]
        rmin=ROI[3]
        
        # Crear una máscara circular
        mask = np.zeros(img.shape[:2], dtype=np.uint8)
        cv2.circle(mask, (x,y), r, 255, -1)
        # Aplicar la máscara a la imagen para seleccionar la ROI circular
        imgROI = cv2.bitwise_and(img, img, mask=mask)
        #Crear el histograma de la ROI
        hist=cv2.calcHist([imgROI],[0],None,[256],[0,255])
        ordhist=sorted(hist, reverse=True)
        maxhist=ordhist[1]
        
        #Fitrado gaussiano para reducir el ruido
        #imgauss = cv2.medianBlur(img, 5)
        imgauss = cv2.GaussianBlur(img, (3, 3), 0)
        
        #Detección de bordes Canny
        imgcanny1 = cv2.Canny(imgauss,umb/2,umb) #Canny de la imagen original con filto
        imgcanny2 = cv2.bitwise_and(imgcanny1, imgcanny1, mask=mask) #ROI de la imagen canny
        imgcanny=imgcanny2[y-r:y+r,x-r:x+r] #acercamento ROI de la imagen canny
        
        if modo=='auto':
            #Umbralizar por Otsu
            umb,imgbn=cv2.threshold(imgauss,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
        elif modo=='manual':
            #Umbralizar por Umbral Manual
            th,imgbn=cv2.threshold(imgauss,umb,255,cv2.THRESH_BINARY)
        """"if polaridad=='blanco':
               contours,hierarchy=cv2.findContours(imgbn,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
        elif polaridad=='negro': #Sacar negativo de la imagen (pues cv2.findContours detecta objetos blancos)
                if modo=='auto':
                    umb,imgbninv=cv2.threshold(imgROI,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
                elif modo=='manual':
                    th,imgbninv=cv2.threshold(imgROI,umb,255,cv2.THRESH_BINARY_INV)
                contours,hierarchy=cv2.findContours(imgbninv,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
        """
        contours,hierarchy=cv2.findContours(imgbn,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
        contours = sorted(contours, key=cv2.contourArea, reverse=True) #Ordena por area, reverse=True se utiliza para ordenar los contornos en orden descendente
        
        imgEdit=img.copy()
        #Dibujar contornos con areas dentro de las tolerancias en azul
        Res=0
        imgR=self.ImgRes(imgEdit,ROI,'red')
        #ROI más grande
        self.MostrarROI(imgcanny)
        ax3.set_title("ROI Bordes Canny",fontsize=10,color='white')
        xc=-1
        yc=-1
        rc=-1
        contcirc=[]
        for i in range (len(contours)):
            #print("Intento1")
            area = cv2.contourArea(contours[i])
            perimetro = cv2.arcLength(contours[i], True)
            # Calcula el circularidad del contorno
            if perimetro!=0:
                circularity = 4 * np.pi * area / perimetro**2
                #print(circularity)
                if (circularity > 0.6):
                    # Encuentra el centro y el radio del círculo
                    (xcont, ycont), rcont = cv2.minEnclosingCircle(contours[i])
                    if (rcont<=tolmax and rcont>=tolmin):
                        contcirc.append([int(xcont),int(ycont),int(rcont)]) 
        if len(contcirc)> 0:
            #print('todos: ',circ)
            radios=[]
            circok=[]
            #print(circok)
            for i in contcirc:
                dist=math.sqrt((x-i[0])**2+(y-i[1])**2)   #Distancia del centro encontrado al centro de la ROI
                if(dist<=rmin):
                    circok.append([i[0],i[1],i[2]])  
                    radios=np.append(radios, i[2])
            #print('circulo ok: ',circok)
            #print('radio ok: ',radios)
            if (len(circok)!= 0):
                indice=np.argmax(radios) #Obtener el circulo de radio mayor
                #print(indice)
                imgR=self.ImgRes(imgEdit,ROI,'green')
                xc=circok[indice][0]
                yc=circok[indice][1]
                rc=circok[indice][2]
                cv2.circle(imgR, (xc, yc), rc, (0, 0, 255), 2) #Dibuja el círculo
                cv2.circle(imgR, (xc, yc), 2, (0, 0, 255), 2) #Dibuja el centro del círculo
                imgcannyc=cv2.cvtColor(imgcanny,cv2.COLOR_GRAY2RGB)
                cv2.circle(imgcannyc, (int(xc-(x-r)), int(yc-(y-r))), rc, (0, 0, 255), 2)
                cv2.circle(imgcannyc, (int(xc-(x-r)), int(yc-(y-r))), 2, (0, 0, 255), 2)
                Res=1        
                #ROI más grande
                self.MostrarROI(imgcannyc)
                ax3.set_title("ROI Bordes Canny",fontsize=10,color='white')
                
        if xc==-1:
            #print("Intento2")
            if modo=='auto':#Sin definir param1
                circ = cv2.HoughCircles(imgauss, cv2.HOUGH_GRADIENT, 1, 3, minRadius=tolmin, maxRadius=tolmax)
            elif modo=='manual':#Definir param1 con umb
                circ = cv2.HoughCircles(imgauss, cv2.HOUGH_GRADIENT, 1, 3, param1=umb, minRadius=tolmin, maxRadius=tolmax)
            if circ is not None and len(circ) > 0:
                circ = np.uint16(np.around(circ))
                #print('todos: ',circ)
                radios=[]
                circok=[]
                for i in circ[0, :]:
                    dist=math.sqrt((x-i[0])**2+(y-i[1])**2)   #Distancia del centro encontrado al centro de la ROI
                    if(dist<=rmin):
                        circok.append([i[0],i[1],i[2]])  
                        radios=np.append(radios, i[2])
                #print('circulo ok: ',circok)
                #print('radio ok: ',radios)
                if (len(circok)!= 0):
                    indice=np.argmax(radios) #Obtener el circulo de radio mayor
                    #print(indice)
                    imgR=self.ImgRes(imgEdit,ROI,'green')
                    xc=circok[indice][0]
                    yc=circok[indice][1]
                    rc=circok[indice][2]
                    cv2.circle(imgR, (xc, yc), rc, (0, 0, 255), 2) #Dibuja el círculo
                    cv2.circle(imgR, (xc, yc), 2, (0, 0, 255), 2) #Dibuja el centro del círculo
                    imgcannyc=cv2.cvtColor(imgcanny,cv2.COLOR_GRAY2RGB)
                    cv2.circle(imgcannyc, (xc-(x-r), yc-(y-r)), rc, (0, 0, 255), 2)
                    cv2.circle(imgcannyc, (xc-(x-r), yc-(y-r)), 2, (0, 0, 255), 2)
                    Res=1        
                    #ROI más grande
                    self.MostrarROI(imgcannyc)
                    ax3.set_title("ROI Bordes Canny",fontsize=10,color='white')

        #Mostrar la imagen con el resultado
        self.MostrarImg(imgR)

        #Graficar histograma
        ax2.clear()
        ax2.plot(hist,'k')
        ax2.axvline(x=umb,color='b')
        ax2.text(umb-10,maxhist+10, umb,fontsize=10,color='b')
        #ax2.axvline(x=int(umb/2),color='b')
        #ax2.text(int(umb/2)-10,maxhist+10, int(umb/2),fontsize=10,color='b')
        ax2.set_title("Histograma de la ROI",fontsize=10)
        ax2.set_xlabel("Nivel de gris")
        ax2.set_ylabel("Pixeles")
        ax2.set_ylim([0,maxhist+10])
        ax2.set_xlim(0,255)
        ax2.grid('on')
        canvas2.draw()
        return Res,xc,yc,rc
    
        
    def DCirculo(self,imgp,ROI):
        global img
        x=ROI[0]
        y=ROI[1]
        r=ROI[2]
        rmin=ROI[3]
        tolmin=rmin
        tolmax=r
        umb=200 #REVISAR
        
        # Crear una máscara circular
        mask = np.zeros(img.shape[:2], dtype=np.uint8)
        cv2.circle(mask, (x,y), r, 255, -1)
        # Aplicar la máscara a la imagen para seleccionar la ROI circular
        imgROI = cv2.bitwise_and(img, img, mask=mask)
        #Crear el histograma de la ROI
        hist=cv2.calcHist([imgROI],[0],None,[256],[0,255])
        ordhist=sorted(hist, reverse=True)
        maxhist=ordhist[1]
        
        #Fitrado gaussiano para reducir el ruido
        imgauss = cv2.GaussianBlur(img, (5, 5), 0)
        
        #Detección de bordes Canny
        imgcanny1 = cv2.Canny(imgauss,umb/2,umb) #Canny de la imagen original con filto
        imgcanny2 = cv2.bitwise_and(imgcanny1, imgcanny1, mask=mask) #ROI de la imagen canny
        imgcanny=imgcanny2[y-r:y+r,x-r:x+r] #acercamento ROI de la imagen canny
        
        #Umbralizar por Otsu
        umb,imgbn=cv2.threshold(imgauss,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
        contours,hierarchy=cv2.findContours(imgbn,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
        contours = sorted(contours, key=cv2.contourArea, reverse=True) #Ordena por area, reverse=True se utiliza para ordenar los contornos en orden descendente
        
        imgR=imgp.copy()
        #Dibujar contornos con area dentro de las tolerancias en azul
        Res=0
        imgcannyc=cv2.cvtColor(imgcanny,cv2.COLOR_GRAY2RGB)
        cv2.circle(imgR,(x,y),r,(255,0,0),3)
        cv2.circle(imgR,(x,y),rmin,(255,0,0),3)
        xc=-1
        yc=-1
        rc=-1
        contcirc=[]
        for i in range (len(contours)):
            area = cv2.contourArea(contours[i])
            perimetro = cv2.arcLength(contours[i], True)
            if perimetro!=0:
                circularity = 4 * np.pi * area / perimetro**2
                if (circularity > 0.6):
                    (xcont, ycont), rcont = cv2.minEnclosingCircle(contours[i])
                    if (rcont<=tolmax and rcont>=tolmin):
                        contcirc.append([int(xcont),int(ycont),int(rcont)]) 
        if len(contcirc)> 0:
            radios=[]
            circok=[]
            #print(circok)
            for i in contcirc:
                dist=math.sqrt((x-i[0])**2+(y-i[1])**2)   #Distancia del centro encontrado al centro de la ROI
                if(dist<=rmin):
                    circok.append([i[0],i[1],i[2]])  
                    radios=np.append(radios, i[2])
            if (len(circok)!= 0):
                indice=np.argmax(radios) #Obtener el circulo de radio mayor
                xc=circok[indice][0]
                yc=circok[indice][1]
                rc=circok[indice][2]
                cv2.circle(imgR, (xc, yc), rc, (0, 0, 255), 2) #Dibuja el círculo
                cv2.circle(imgR, (xc, yc), 2, (0, 0, 255), 2) #Dibuja el centro del círculo
                cv2.circle(imgcannyc, (int(xc-(x-r)), int(yc-(y-r))), rc, (0, 0, 255), 2)
                cv2.circle(imgcannyc, (int(xc-(x-r)), int(yc-(y-r))), 2, (0, 0, 255), 2)
                cv2.circle(imgR,(x,y),r,(0,255,0),3)
                cv2.circle(imgR,(x,y),rmin,(0,255,0),3)
                Res=1        
                
        if xc==-1:
            circ = cv2.HoughCircles(imgauss, cv2.HOUGH_GRADIENT, 1, 5, minRadius=tolmin, maxRadius=tolmax)
            if circ is not None and len(circ) > 0:
                circ = np.uint16(np.around(circ))
                radios=[]
                circok=[]
                for i in circ[0, :]:
                    dist=math.sqrt((x-i[0])**2+(y-i[1])**2)   #Distancia del centro encontrado al centro de la ROI
                    if(dist<=rmin):
                        circok.append([i[0],i[1],i[2]])  
                        radios=np.append(radios, i[2])
                if (len(circok)!= 0):
                    indice=np.argmax(radios) #Obtener el circulo de radio mayor
                    xc=circok[indice][0]
                    yc=circok[indice][1]
                    rc=circok[indice][2]
                    cv2.circle(imgR, (xc, yc), rc, (0, 0, 255), 2) #Dibuja el círculo
                    cv2.circle(imgR, (xc, yc), 2, (0, 0, 255), 2) #Dibuja el centro del círculo
                    cv2.circle(imgcannyc, (xc-(x-r), yc-(y-r)), rc, (0, 0, 255), 2)
                    cv2.circle(imgcannyc, (xc-(x-r), yc-(y-r)), 2, (0, 0, 255), 2)
                    cv2.circle(imgR,(x,y),r,(0,255,0),3)
                    cv2.circle(imgR,(x,y),rmin,(0,255,0),3)
                    Res=1        
        return Res,xc,yc,rc,imgR,imgcannyc,hist,
    
    def PtsROILinea(self,ROI):
        xi=ROI[0]
        yi=ROI[1]
        xf=ROI[2]
        yf=ROI[3]
        mw=ROI[4]/2
        if (xi==xf): #Linea vertical m=infinito
            x1=xi-mw
            y1=yi
            x2=xi+mw
            y2=yi
            x4=xf+mw
            y4=yf
            x3=xf-mw
            y3=yf
        else:
            mROI=(yf-yi)/(xf-xi) 
            if (mROI==0): #Linea horizontal m=0
                x1=xi
                y1=yi+mw
                x2=xi
                y2=yi-mw
                x4=xf
                y4=yf-mw
                x3=xf
                y3=yf+mw
                a1=xi
                a2=xf
            else:
                x, y = symbols('x y')
                mperp=-1/mROI
                b1=-(mperp*xi)+yi
                b2=-(mperp*xf)+yf
                Ec1=y-mperp*x-b1
                Ec2=(xi-x)**2+(yi-y)**2-mw**2
                xy1=solve([Ec1,Ec2],[x,y])
                x1=xy1[0][0]
                y1=xy1[0][1]
                x2=xy1[1][0]
                y2=xy1[1][1]
                Ec3=y-mperp*x-b2
                Ec4=(xf-x)**2+(yf-y)**2-mw**2
                xy2=solve([Ec3,Ec4],[x,y])
                x3=xy2[0][0]
                y3=xy2[0][1]
                x4=xy2[1][0]
                y4=xy2[1][1]
        xs=np.array([x1,x2,x3,x4])
        ys=np.array([y1,y2,y3,y4])
        pts=np.array([[x1,y1],[x2,y2],[x4,y4],[x3,y3]],np.int32)
        return pts,xs,ys
    
    def Linea(self,img,ROI,tol=10,modo='auto',ubajo=127,ualto=255): 
        xi=ROI[0]
        yi=ROI[1]
        xf=ROI[2]
        yf=ROI[3]
        mw=ROI[4]/2
        if (xi==xf): #Linea vertical m=infinito
            x1=xi-mw
            y1=yi
            x2=xi+mw
            y2=yi
            x4=xf+mw
            y4=yf
            x3=xf-mw
            y3=yf
            mperp=0
            b1=yi
            b2=yf
            mROI=[]
            angROI=90 #ángulo de la recta dada
        else:
            mROI=(yf-yi)/(xf-xi) 
            angROI=math.degrees(math.atan(mROI)) #ángulo de la recta dada
            #print('mROI',m)
            if (mROI==0): #Linea horizontal m=0
                x1=xi
                y1=yi+mw
                x2=xi
                y2=yi-mw
                x4=xf
                y4=yf-mw
                x3=xf
                y3=yf+mw
                a1=xi
                a2=xf
            else:
                x, y = symbols('x y')
                #b=-(mROI*xi)+yi
                mperp=-1/mROI
                b1=-(mperp*xi)+yi
                b2=-(mperp*xf)+yf
                Ec1=y-mperp*x-b1
                Ec2=(xi-x)**2+(yi-y)**2-mw**2
                xy1=solve([Ec1,Ec2],[x,y])
                #x1=np.amin(xy1[0])
                x1=xy1[0][0]
                y1=xy1[0][1]
                x2=xy1[1][0]
                y2=xy1[1][1]
                Ec3=y-mperp*x-b2
                Ec4=(xf-x)**2+(yf-y)**2-mw**2
                xy2=solve([Ec3,Ec4],[x,y])
                x3=xy2[0][0]
                y3=xy2[0][1]
                x4=xy2[1][0]
                y4=xy2[1][1]
        xs=np.array([x1,x2,x3,x4])
        ys=np.array([y1,y2,y3,y4])
        pts=np.array([[x1,y1],[x2,y2],[x4,y4],[x3,y3]],np.int32)

        # Crear una máscara rectangular rotada
        mask = np.zeros(img.shape[:2], dtype=np.uint8)
        cv2.fillPoly(mask, [pts],255)
        # Aplicar la máscara a la imagen para seleccionar la ROI
        imgROI = cv2.bitwise_and(img, img, mask=mask)
        #Crear el histograma de la ROI
        hist=cv2.calcHist([imgROI],[0],None,[256],[0,255])
        ordhist=sorted(hist, reverse=True)
        maxhist=ordhist[1]

        #Fitrado gaussiano para reducir el ruido
        imgauss = cv2.GaussianBlur(img, (3, 3), 0)

        #Detección de bordes Canny
        if modo=='auto':
            imgcanny = cv2.Canny(imgauss,70,200) ##Revisar umbrales
        elif modo=='manual':
            imgcanny = cv2.Canny(imgauss,ubajo,ualto)
        imgcannyROI = cv2.bitwise_and(imgcanny, imgcanny, mask=mask) #ROI de la imagen canny
        
        # Detección de linea utilizando la transformada de Hough
        lin = cv2.HoughLines(imgcanny,1,np.pi/180,50) #Revisar threshold(acumulado mínimo que debe obtener para que se le considere como una línea)
        
        imgEdit=img.copy()
        #Dibujar linea dentro de las tolerancias en azul
        Res=0
        imgR=self.ImgRes(imgEdit,ROI,'red')
        #ROI más grande
        self.MostrarROI(imgcannyROI[int(np.min(ys)):int(np.max(ys)),int(np.min(xs)):int(np.max(xs))])
        ax3.set_title("ROI Bordes Canny",fontsize=10,color='white')
        x1lin=-1
        y1lin=-1
        x2lin=-1
        y2lin=-1
        if lin is not None and len(lin) > 0:
            distancias=[]
            lineas=[]
            for i in lin:
                rho,theta=i[0]
                cost=np.cos(theta)
                x0=cost*rho
                sint=np.sin(theta)
                y0=sint*rho
                x0=round(x0)
                y0=round(y0)
                #print('y0',y0)
                #print('cos',cost)
                #print('sin',sint)
                if (sint==0): #La linea encontrada es vertical m=infinito
                    ang=90
                    a=x0 #Linea encontrada vertical x=a
                else:
                    m=-cost/sint #Pendiente de la linea encontrada
                    #print('mlinea',m)
                    ang=math.degrees(math.atan(m)) #Angulo de inclinación de la linea encontrada
                    #ang=math.degrees(theta-(np.pi/2))
                    #y=m(x-x0)+y0 --> y=mx-mx0+y0 --> y=mx+b --> b=-mx0+y0 
                    b=-(m*x0)+y0 #Ordenada al origen de la linea encontrada
                    #print('b',b)         
                if (ang<angROI+tol and ang>angROI-tol):
                    if (mROI==0): #Lineas perpendiculares verticales x=a
                        if (sint==0): #Linea encontrada vertical x=a
                            A1=np.array([[1,0],[1,0]])
                            B1=np.array([a1,a])
                            A2=np.array([[1,0],[1,0]])
                            B2=np.array([a2,a])
                        else:
                            A1=np.array([[1,0],[-m,1]])
                            B1=np.array([a1,b])
                            A2=np.array([[1,0],[-m,1]])
                            B2=np.array([a2,b])
                    else:#mx+y=b
                        if (sint==0): #Linea encontrada vertical x=a
                            A1=np.array([[-mperp,1],[1,0]])
                            B1=np.array([b1,a])
                            A2=np.array([[-mperp,1],[1,0]])
                            B2=np.array([b2,a])
                        else:
                            A1=np.array([[-mperp,1],[-m,1]])
                            B1=np.array([b1,b])
                            A2=np.array([[-mperp,1],[-m,1]])
                            B2=np.array([b2,b])
                    coor1 = np.linalg.solve(A1, B1)
                    coor2 = np.linalg.solve(A2, B2)
                    #print('coordenadas 1',coor1)
                    #print('coordenadas 2',coor2)
                    xenc=np.array([coor1[0],coor2[0]])
                    yenc=np.array([coor1[1],coor2[1]])
                    if (np.max(xenc)<np.max(xs) and np.min(xenc)>np.min(xs) and np.max(yenc)<np.max(ys) and np.min(yenc)>np.min(ys)):
                        #cv2.line(imlin1, (round(coor1[0]),round(coor1[1])), (round(coin2[0]),round(coin2[1])), (0, 0, 255), 3, cv2.LINE_AA)
                        dist=math.sqrt((np.min(xenc)-np.min(xs))**2+(np.min(yenc)-np.min(ys))**2)
                        #print('distancia',dist)
                        distancias=np.append(distancias,dist)
                        lineas.append([coor1[0],coor1[1],coor2[0],coor2[1]])                      
            #print(lineas)
            if (len(lineas)!= 0):
                indice=np.argmin(distancias) #Obtener la linea más cercana a la ROI  
                imgR=self.ImgRes(imgEdit,ROI,'green')
                x1lin=round(lineas[indice][0])
                y1lin=round(lineas[indice][1])
                x2lin=round(lineas[indice][2])
                y2lin=round(lineas[indice][3])
                cv2.line(imgR, (x1lin,y1lin),(x2lin,y2lin), (0, 0, 255), 2, cv2.LINE_AA)
                imgcannyROIl=cv2.cvtColor(imgcannyROI,cv2.COLOR_GRAY2RGB)
                cv2.line(imgcannyROIl, (x1lin,y1lin),(x2lin,y2lin), (0, 0, 255), 2, cv2.LINE_AA)
                Res=1 
                #ROI más grande
                self.MostrarROI(imgcannyROIl[int(np.min(ys)):int(np.max(ys)),int(np.min(xs)):int(np.max(xs))])
                ax3.set_title("ROI Bordes Canny",fontsize=10,color='white')
                
        #Mostrar la imagen con el resultado
        self.MostrarImg(imgR)
        
        #Graficar histograma
        ax2.clear()
        ax2.plot(hist,'k')
        ax2.axvline(x=ubajo,color='b')
        ax2.text(ubajo-10,maxhist+10,ubajo,fontsize=10,color='b')
        ax2.axvline(x=ualto ,color='b')
        ax2.text(ualto-10,maxhist+10,ualto,fontsize=10,color='b')
        ax2.set_title("Histograma de la ROI",fontsize=10)
        ax2.set_xlabel("Nivel de gris")
        ax2.set_ylabel("Pixeles")
        ax2.set_ylim([0,maxhist+10])
        ax2.set_xlim(0,255)
        ax2.grid('on')
        canvas2.draw()
        return Res,x1lin,y1lin,x2lin,y2lin
    
    
    def DLinea(self,imgp,ROI): 
        global img
        xi=ROI[0]
        yi=ROI[1]
        xf=ROI[2]
        yf=ROI[3]
        mw=ROI[4]/2
        tol=10
        ubajo=70
        ualto=200 #REVISAR
        if (xi==xf): #Linea vertical m=infinito
            x1=xi-mw
            y1=yi
            x2=xi+mw
            y2=yi
            x4=xf+mw
            y4=yf
            x3=xf-mw
            y3=yf
            mperp=0
            b1=yi
            b2=yf
            mROI=[]
            angROI=90 #ángulo de la recta dada
        else:
            mROI=(yf-yi)/(xf-xi) 
            angROI=math.degrees(math.atan(mROI)) #ángulo de la recta dada
            #print('mROI',m)
            if (mROI==0): #Linea horizontal m=0
                x1=xi
                y1=yi+mw
                x2=xi
                y2=yi-mw
                x4=xf
                y4=yf-mw
                x3=xf
                y3=yf+mw
                a1=xi
                a2=xf
            else:
                x, y = symbols('x y')
                #b=-(mROI*xi)+yi
                mperp=-1/mROI
                b1=-(mperp*xi)+yi
                b2=-(mperp*xf)+yf
                Ec1=y-mperp*x-b1
                Ec2=(xi-x)**2+(yi-y)**2-mw**2
                xy1=solve([Ec1,Ec2],[x,y])
                #print('R1',xy1)
                #x1=np.amin(xy1[0])
                x1=xy1[0][0]
                y1=xy1[0][1]
                x2=xy1[1][0]
                y2=xy1[1][1]
                Ec3=y-mperp*x-b2
                Ec4=(xf-x)**2+(yf-y)**2-mw**2
                xy2=solve([Ec3,Ec4],[x,y])
                #print('R2',xy2)
                x3=xy2[0][0]
                y3=xy2[0][1]
                x4=xy2[1][0]
                y4=xy2[1][1]
        xs=np.array([x1,x2,x3,x4])
        ys=np.array([y1,y2,y3,y4])
        pts=np.array([[x1,y1],[x2,y2],[x4,y4],[x3,y3]],np.int32)

        # Crear una máscara rectangular rotada
        mask = np.zeros(img.shape[:2], dtype=np.uint8)
        cv2.fillPoly(mask, [pts],255)
        # Aplicar la máscara a la imagen para seleccionar la ROI
        imgROI = cv2.bitwise_and(img, img, mask=mask)
        #Crear el histograma de la ROI
        hist=cv2.calcHist([imgROI],[0],None,[256],[0,255])
        ordhist=sorted(hist, reverse=True)
        maxhist=ordhist[1]

        #Fitrado gaussiano para reducir el ruido
        imgauss = cv2.GaussianBlur(img, (3, 3), 0)

        #Detección de bordes Canny
        imgcanny = cv2.Canny(imgauss,70,200) ##Revisar umbrales
        imgcannyROI = cv2.bitwise_and(imgcanny, imgcanny, mask=mask) #ROI de la imagen canny
        
        # Detección de linea utilizando la transformada de Hough
        lin = cv2.HoughLines(imgcanny,1,np.pi/180,50) #Revisar threshold(acumulado mínimo que debe obtener para que se le considere como una línea)
        
        imgR=imgp.copy()
        #Dibujar linea dentro de las tolerancias en azul
        Res=0
        imgcannyROIl=cv2.cvtColor(imgcannyROI,cv2.COLOR_GRAY2RGB)
        cv2.polylines(imgR,[pts],True,(255,0,0),3)
        x1lin=-1
        y1lin=-1
        x2lin=-1
        y2lin=-1
        if lin is not None and len(lin) > 0:
            distancias=[]
            lineas=[]
            for i in lin:
                rho,theta=i[0]
                cost=np.cos(theta)
                x0=cost*rho
                sint=np.sin(theta)
                y0=sint*rho
                x0=round(x0)
                y0=round(y0)
                if (sint==0): #La linea encontrada es vertical m=infinito
                    ang=90
                    a=x0 #Linea encontrada vertical x=a
                else:
                    m=-cost/sint #Pendiente de la linea encontrada
                    ang=math.degrees(math.atan(m)) #Angulo de inclinación de la linea encontrada
                    #y=m(x-x0)+y0 --> y=mx-mx0+y0 --> y=mx+b --> b=-mx0+y0 
                    b=-(m*x0)+y0 #Ordenada al origen de la linea encontrada    
                if (ang<angROI+tol and ang>angROI-tol):
                    if (mROI==0): #Lineas perpendiculares verticales x=a
                        if (sint==0): #Linea encontrada vertical x=a
                            A1=np.array([[1,0],[1,0]])
                            B1=np.array([a1,a])
                            A2=np.array([[1,0],[1,0]])
                            B2=np.array([a2,a])
                        else:
                            A1=np.array([[1,0],[-m,1]])
                            B1=np.array([a1,b])
                            A2=np.array([[1,0],[-m,1]])
                            B2=np.array([a2,b])
                    else:#mx+y=b
                        if (sint==0): #Linea encontrada vertical x=a
                            A1=np.array([[-mperp,1],[1,0]])
                            B1=np.array([b1,a])
                            A2=np.array([[-mperp,1],[1,0]])
                            B2=np.array([b2,a])
                        else:
                            A1=np.array([[-mperp,1],[-m,1]])
                            B1=np.array([b1,b])
                            A2=np.array([[-mperp,1],[-m,1]])
                            B2=np.array([b2,b])
                    coor1 = np.linalg.solve(A1, B1)
                    coor2 = np.linalg.solve(A2, B2)
                    xenc=np.array([coor1[0],coor2[0]])
                    yenc=np.array([coor1[1],coor2[1]])
                    if (np.max(xenc)<np.max(xs) and np.min(xenc)>np.min(xs) and np.max(yenc)<np.max(ys) and np.min(yenc)>np.min(ys)):
                        dist=math.sqrt((np.min(xenc)-np.min(xs))**2+(np.min(yenc)-np.min(ys))**2)
                        distancias=np.append(distancias,dist)
                        lineas.append([coor1[0],coor1[1],coor2[0],coor2[1]])                      
            if (len(lineas)!= 0):
                indice=np.argmin(distancias) #Obtener la linea más cercana a la ROI  
                x1lin=round(lineas[indice][0])
                y1lin=round(lineas[indice][1])
                x2lin=round(lineas[indice][2])
                y2lin=round(lineas[indice][3])
                cv2.line(imgR, (x1lin,y1lin),(x2lin,y2lin), (0, 0, 255), 2, cv2.LINE_AA)
                cv2.line(imgcannyROIl, (x1lin,y1lin),(x2lin,y2lin), (0, 0, 255), 2, cv2.LINE_AA)
                cv2.polylines(imgR,[pts],True,(0,255,0),3)
                Res=1
        imgcannyROIs=imgcannyROIl[int(np.min(ys)):int(np.max(ys)),int(np.min(xs)):int(np.max(xs))]
                
        return Res,x1lin,y1lin,x2lin,y2lin,imgR,imgcannyROIs,hist
       
    def ImgRes(self,img,ROI,color):
        global Opciones,Herramienta,Medicion
        if (img[img==0].size!=img.size):
            choose=Opciones.get() 
            imgR=img.copy()
            #Segun el color de marco elegido
            if color=='red':
                c=(255,0,0)
            elif color=='green':
                c=(0,255,0)
            elif color=='blue':
                c=(0,0,255)
            if(choose==1):
                #Convertir la imagen a RGB
                imgR=cv2.cvtColor(img,cv2.COLOR_GRAY2RGB)
                tool=Herramienta.get()
                x=ROI[0]
                y=ROI[1]
                w=ROI[2]
                h=ROI[3]          
                if (tool<4):
                    cv2.rectangle(imgR,(x,y),(x+w,y+h),c,3)
                elif (tool==4):
                    cv2.circle(imgR,(x,y),w,c,3)
                    cv2.circle(imgR,(x,y),h,c,3)
                elif (tool==5):
                    pts,ptsx,ptsy=self.PtsROILinea(ROI)
                    cv2.polylines(imgR,[pts],True,c,3)
            if (choose==2):
                med=Medicion.get()
                ROI1=ROI[0]
                ROI2=ROI[1]
                x1=ROI1[0]
                y1=ROI1[1]
                w1=ROI1[2]
                h1=ROI1[3]
                x2=ROI2[0]
                y2=ROI2[1]
                w2=ROI2[2]
                h2=ROI2[3]
                if(med==1 or med==5):
                    cv2.circle(imgR,(x1,y1),w1,c,3)
                    cv2.circle(imgR,(x1,y1),h1,c,3)
                    cv2.circle(imgR,(x2,y2),w2,c,3)
                    cv2.circle(imgR,(x2,y2),h2,c,3)
                elif(med==2):
                    cv2.circle(imgR,(x1,y1),w1,c,3)
                    cv2.circle(imgR,(x1,y1),h1,c,3)
                    pts,ptsx,ptsy=self.PtsROILinea(ROI2)
                    cv2.polylines(imgR,[pts],True,c,3)
                elif(med==3 or med==4):
                    pts1,ptsx1,ptsy1=self.PtsROILinea(ROI1)
                    pts2,ptsx2,ptsy2=self.PtsROILinea(ROI2)
                    cv2.polylines(imgR,[pts1],True,c,3)
                    cv2.polylines(imgR,[pts2],True,c,3)
        else:
            tk.messagebox.showwarning(message="Carga una imagen para trabajar")
        return imgR    
        
    def ROIentry(self,event):
        self.ROI()
        
    def ROI(self):
        global Opciones,img,Figura,ROI1,ROI2,ax2,canvas2,px,py,ancho,alto,mw,labROIs,labaROI,Herramienta,Medicion
        if (img[img==0].size!=img.size):
            choose=Opciones.get() 
            x=px.get()
            y=py.get()
            w=ancho.get()
            h=alto.get()
            mancho=mw.get()
            
            if (x>img.shape[1]):
                x=img.shape[1]
                px.set(x)
            elif (x<0):
                x=0
                px.set(x)
            if (y>img.shape[0]):
                y=img.shape[0]
                py.set(y)
            elif (y<0):
                y=0
                px.set(y)
                    
            if(choose==1):
                ROI=[x,y,w,h,mancho]
                tool=Herramienta.get()
                #print(tool)
                if (tool<4):
                    if ((x+w)>img.shape[1]):
                        w=img.shape[1]-x
                        ancho.set(w)
                    if ((y+h)>img.shape[0]):
                        h=img.shape[0]-y    
                        alto.set(h)
                    ROI=[x,y,w,h,mancho]
                    labROIs.config(text=f"ROI=( {x}, {y}, {w}, {h})")
                elif (tool==4):
                    r=w    
                    if ((x-r)<0):
                        w=x
                    elif ((x+r)>img.shape[1]):
                        w=img.shape[1]-x
                    if((y-r)<0):
                        w=y
                    elif ((y+r)>img.shape[0]):
                        w=img.shape[0]-y
                    ancho.set(int(w)) 
                    ROI=[x,y,w,h,mancho]
                    labROIs.config(text=f"ROI=c:({x}, {y}) rmin:{h} rmax:{w}")
                    # Crear una máscara circular
                    mask = np.zeros(img.shape[:2], dtype=np.uint8)
                    cv2.circle(mask, (x,y), r, 255, -1)
                elif (tool==5):
                    if (w>img.shape[1]):
                        w=img.shape[1]
                        ancho.set(w)
                    elif (w<0):
                        w=0
                        ancho.set(w)
                    if (h>img.shape[0]):
                        h=img.shape[0]
                        alto.set(h)
                    elif (h<0):
                        h=0
                        alto.set(h)
                    ROI=[x,y,w,h,mancho]
                    labROIs.config(text=f"ROI=P1:({x}, {y}) P2:({w}, {h}) w:{mancho}")
                    pts,ptsx,ptsy=self.PtsROILinea(ROI)
                    # Crear una máscara rectangular rotada
                    mask = np.zeros(img.shape[:2], dtype=np.uint8)
                    cv2.fillPoly(mask, [pts],255)

                if (tool==4 or tool==5):
                    # Aplicar la máscara a la imagen para seleccionar la ROI
                    imgROI = cv2.bitwise_and(img, img, mask=mask)
                    #Crear el histograma de la ROI
                    hist=cv2.calcHist([imgROI],[0],None,[256],[0,255])
                    ordhist=sorted(hist, reverse=True)
                    maxhist=ordhist[1]
                     #Graficar histograma
                    ax2.clear()
                    ax2.plot(hist,'k')
                    ax2.set_title("Histograma de la ROI",fontsize=10)
                    ax2.set_xlabel("Nivel de gris")
                    ax2.set_ylabel("Pixeles")
                    ax2.set_ylim([0,maxhist+10])
                    ax2.set_xlim(0,255)
                    ax2.grid('on')
                    canvas2.draw()
                    if(tool==4):
                        imgROI=imgROI[y-r:y+r,x-r:x+r]
                    elif(tool==5):
                        imgROI=imgROI[int(np.min(ptsy)):int(np.max(ptsy)),int(np.min(ptsx)):int(np.max(ptsx))]    
                else:
                    #Extraer la ROI
                    imgROI=img[y:y+h,x:x+w]
                    #Crear el histograma de la ROI
                    hist=cv2.calcHist([imgROI],[0],None,[256],[0,256])
                    #Graficar histograma
                    ax2.clear()
                    ax2.plot(hist,'k')
                    ax2.set_title("Histograma de la ROI",fontsize=10)
                    ax2.set_xlabel("Nivel de gris")
                    ax2.set_ylabel("Pixeles")
                    ax2.set_xlim(0,255)
                    ax2.grid('on')
                    canvas2.draw()
                self.Tolerancias()
            if (choose==2):
                med=Medicion.get()
                #print(med)
                figure=Figura.get()
                if(med==1 or med==5):
                    r=w    
                    if ((x-r)<0):
                        w=x
                    elif ((x+r)>img.shape[1]):
                        w=img.shape[1]-x
                    if((y-r)<0):
                        w=y
                    elif ((y+r)>img.shape[0]):
                        w=img.shape[0]-y
                    ancho.set(int(w)) 
                    ROI=[x,y,w,h,mancho]
                    labROIs.config(text=f"ROI=c:({x}, {y}) rmin:{h} rmax:{w}")
                    # Crear una máscara circular
                    mask = np.zeros(img.shape[:2], dtype=np.uint8)
                    cv2.circle(mask, (x,y), r, 255, -1)
                elif(med==2):
                    if (figure==1):
                        r=w    
                        if ((x-r)<0):
                            w=x
                        elif ((x+r)>img.shape[1]):
                            w=img.shape[1]-x
                        if((y-r)<0):
                            w=y
                        elif ((y+r)>img.shape[0]):
                            w=img.shape[0]-y
                        ancho.set(int(w)) 
                        labROIs.config(text=f"ROI=c:({x}, {y}) rmin:{h} rmax:{w}")
                        ROI=[x,y,w,h,mancho]
                        # Crear una máscara circular
                        mask = np.zeros(img.shape[:2], dtype=np.uint8)
                        cv2.circle(mask, (x,y), r, 255, -1)        
                    elif (figure==2):
                        if (w>img.shape[1]):
                            w=img.shape[1]
                            ancho.set(w)
                        elif (w<0):
                            w=0
                            ancho.set(w)
                        if (h>img.shape[0]):
                            h=img.shape[0]
                            alto.set(h)
                        elif (h<0):
                            h=0
                            alto.set(h)
                        ROI=[x,y,w,h,mancho]
                        labROIs.config(text=f"ROI=P1:({x}, {y}) P2:({w}, {h}) w:{mancho}")
                        pts,ptsx,ptsy=self.PtsROILinea(ROI)
                        # Crear una máscara rectangular rotada
                        mask = np.zeros(img.shape[:2], dtype=np.uint8)
                        cv2.fillPoly(mask, [pts],255)
                elif(med==3 or med==4):
                    if (w>img.shape[1]):
                        w=img.shape[1]
                        ancho.set(w)
                    elif (w<0):
                        w=0
                        ancho.set(w)
                    if (h>img.shape[0]):
                        h=img.shape[0]
                        alto.set(h)
                    elif (h<0):
                        h=0
                        alto.set(h)
                    labROIs.config(text=f"ROI=P1:({x}, {y}) P2:({w}, {h}) w:{mancho}")
                    ROI=[x,y,w,h,mancho]
                    pts,ptsx,ptsy=self.PtsROILinea(ROI)
                    # Crear una máscara rectangular rotada
                    mask = np.zeros(img.shape[:2], dtype=np.uint8)
                    cv2.fillPoly(mask, [pts],255)
                
                # Aplicar la máscara a la imagen para seleccionar la ROI
                imgROI = cv2.bitwise_and(img, img, mask=mask)
                #Crear el histograma de la ROI
                hist=cv2.calcHist([imgROI],[0],None,[256],[0,255])
                ordhist=sorted(hist, reverse=True)
                maxhist=ordhist[1]
                #Graficar histograma
                ax2.clear()
                ax2.plot(hist,'k')
                ax2.set_title("Histograma de la ROI",fontsize=10)
                ax2.set_xlabel("Nivel de gris")
                ax2.set_ylabel("Pixeles")
                ax2.set_ylim([0,maxhist+10])
                ax2.set_xlim(0,255)
                ax2.grid('on')
                canvas2.draw()
                
                if(med==1 or med==5):
                    imgROI=imgROI[y-r:y+r,x-r:x+r]
                elif(med==2):
                    if (figure==1):
                        imgROI=imgROI[y-r:y+r,x-r:x+r]      
                    elif (figure==2):
                         imgROI=imgROI[int(np.min(ptsy)):int(np.max(ptsy)),int(np.min(ptsx)):int(np.max(ptsx))]    
                elif(med==3 or med==4):
                    imgROI=imgROI[int(np.min(ptsy)):int(np.max(ptsy)),int(np.min(ptsx)):int(np.max(ptsx))]    
                
                if (figure==1):
                    ROI1=[px.get(),py.get(),ancho.get(),alto.get(),mw.get()]
                elif (figure==2):
                    ROI2=[px.get(),py.get(),ancho.get(),alto.get(),mw.get()]
                
                ROI=[ROI1,ROI2]
                
            labaROI.config(text=f"{w*h} píxeles")
            #Visualización de ROI enamarcada en azul y acercamiento
            imgR=self.ImgRes(img,ROI,'blue')
            self.MostrarImg(imgR)
            self.MostrarROI(imgROI)
        else:
            tk.messagebox.showwarning(message="Carga una imagen para trabajar")
                     
    def MostrarROI(self,img):
        ax3.clear()
        ax3.imshow(img,'gray')
        ax3.axis('off')
        ax3.set_title("ROI",fontsize=10,color='white')
        canvas3.draw() 
        
    def motion(self,event):
        global img,labpos
        #print(img.shape)
        sx=math.floor((640-img.shape[1])/2) #math.ceil redondeado hacia arriba
        sy=math.floor((480-img.shape[0])/2) #math.floor redondeado hacia abajo
        x=event.x-sx
        y=event.y-sy
        #print("x y: (%s %s)" % (sx, sy))
        #print("Mouse position: (%s %s)" % (event.x, event.y))
        #print("Image position: (%s %s)" % (x,y))
        labpos.config(text=f"x:{x}, y:{y}")
        
    def ROIxy(self,event):
        global img,px,py
        sx=math.floor((640-img.shape[1])/2) #math.ceil redondeado hacia arriba
        sy=math.floor((480-img.shape[0])/2)
        x=event.x-sx
        y=event.y-sy
        
        if (x<0):
            x=0
        elif (x>img.shape[1]):
            x=img.shape[1]
        if(y<0):
            y=0
        elif (y>img.shape[0]):
            y=img.shape[0]
        px.set(x)
        py.set(y)
         
    def ROIwh(self,event):
        global img,px,py,ancho,alto,mw,Opciones,Herramienta,Medicion,Figura
        choose=Opciones.get() 
        sx=math.floor((640-img.shape[1])/2) #math.ceil redondeado hacia arriba
        sy=math.floor((480-img.shape[0])/2)
        x=event.x-sx
        y=event.y-sy
        if (x<0):
            x=0
        elif (x>img.shape[1]):
            x=img.shape[1]
        if(y<0):
            y=0
        elif (y>img.shape[0]):
            y=img.shape[0]
        
        x1=px.get()
        y1=py.get()
        
        tool=Herramienta.get() 
        if(choose==1):
            if (tool<4):
                w=x-x1
                h=y-y1
                if (w!=0):
                    ancho.set(w)
                else:
                    ancho.set(1)
                if (h!=0):
                    alto.set(h)
                else:
                    alto.set(1)
            elif (tool==4):
                r=math.sqrt((x-x1)**2+(y-y1)**2)    
                if ((x1-r)<0):
                    r=x1
                elif ((x1+r)>img.shape[1]):
                    r=img.shape[1]-x1
                if((y1-r)<0):
                    r=y1
                elif ((y1+r)>img.shape[0]):
                    r=img.shape[0]-y1
                ancho.set(int(r)) 
            elif (tool==5):
                ancho.set(x)
                alto.set(y)
            self.Tolerancias()
        if (choose==2):
            med=Medicion.get()
            figure=Figura.get()
            if(med==1 or med==5):
                r=math.sqrt((x-x1)**2+(y-y1)**2)    
                if ((x1-r)<0):
                    r=x1
                elif ((x1+r)>img.shape[1]):
                    r=img.shape[1]-x1
                if((y1-r)<0):
                    r=y1
                elif ((y1+r)>img.shape[0]):
                    r=img.shape[0]-y1
                ancho.set(int(r))        
            elif(med==2):
                if (figure==1):
                    r=math.sqrt((x-x1)**2+(y-y1)**2)    
                    if ((x1-r)<0):
                        r=x1
                    elif ((x1+r)>img.shape[1]):
                        r=img.shape[1]-x1
                    if((y1-r)<0):
                        r=y1
                    elif ((y1+r)>img.shape[0]):
                        r=img.shape[0]-y1
                    ancho.set(int(r))              
                elif (figure==2):
                    ancho.set(x)
                    alto.set(y) 
            elif(med==3 or med==4):
                ancho.set(x)
                alto.set(y)     
                
            if (figure==1):
                ROI1=[px.get(),py.get(),ancho.get(),alto.get(),mw.get()]
            elif (figure==2):
                ROI2=[px.get(),py.get(),ancho.get(),alto.get(),mw.get()]
        if (choose==1 and tool==0):
            self.ROI()
        else:
            self.Vision()
    
    def ROIrmin(self,event):
        global img,px,py,ancho,alto,mw,Opciones,Herramienta,Medicion,Figura
        choose=Opciones.get() 
        sx=math.floor((640-img.shape[1])/2) #math.ceil redondeado hacia arriba
        sy=math.floor((480-img.shape[0])/2)
        x=event.x-sx
        y=event.y-sy
        if (x<0):
            x=0
        elif (x>img.shape[1]):
            x=img.shape[1]
        if(y<0):
            y=0
        elif (y>img.shape[0]):
            y=img.shape[0]
        
        x1=px.get()
        y1=py.get()      
        rmax=ancho.get()
        x2=ancho.get()
        y2=alto.get()
        
        if(choose==1):
            tool=Herramienta.get()    
            if (tool==4):
                r=math.sqrt((x-x1)**2+(y-y1)**2)  
                if (r>rmax):
                    r=rmax
                alto.set(int(r)) 
            if (tool==5): #Calcular la distancia de la linea al click 
                #y=m(x-x0)+y0 --> y=mx-mx0+y0 --> mx-y-mx0+y0=0
                m=(y2-y1)/(x2-x1) 
                #Distancia de a recta Ax+By+C=0 al punto (x0,y0)
                #Ax+By+C=0--> mx-y-mx0+y0=0 --> A=m B=-1 C=-mx0+y0=0
                #d=|(Ax0+By0+C)/sqrt(A**2+B**2)|
                mancho=abs(((m*x)+(-1*y)+(-(m*x1)+y1))/(math.sqrt((m)**2+(-1)**2)))
                #print(mancho)
                mw.set(int(mancho*2)) 
            self.Tolerancias()
        if (choose==2):
            med=Medicion.get()
            figure=Figura.get()
            if(med==1 or med==5):
                r=math.sqrt((x-x1)**2+(y-y1)**2)  
                if (r>rmax):
                    r=rmax
                alto.set(int(r))    
            elif(med==2):
                if (figure==1):
                    r=math.sqrt((x-x1)**2+(y-y1)**2)  
                    if (r>rmax):
                        r=rmax
                    alto.set(int(r))      
                elif (figure==2):
                    m=(y2-y1)/(x2-x1) 
                    mancho=abs(((m*x)+(-1*y)+(-(m*x1)+y1))/(math.sqrt((m)**2+(-1)**2)))
                    mw.set(int(mancho*2)) 
            elif(med==3 or med==4):
                m=(y2-y1)/(x2-x1) 
                mancho=abs(((m*x)+(-1*y)+(-(m*x1)+y1))/(math.sqrt((m)**2+(-1)**2)))
                mw.set(int(mancho*2)) 
            
            if (figure==1):
                ROI1=[px.get(),py.get(),ancho.get(),alto.get(),mw.get()]
            elif (figure==2):
                ROI2=[px.get(),py.get(),ancho.get(),alto.get(),mw.get()]
        self.Vision()
        

    def guardar_datos(self):
        global entry_matriz,top,centimetros,pixeles,Ls
        # Crea una nueva matriz con los datos ingresados en los Entry
        centimetros = []
        for fila_entry in entry_matriz:
            fila_nueva_matriz = []
            for entry in fila_entry:
                fila_nueva_matriz.append(int(entry.get()))
            centimetros.append(fila_nueva_matriz)
        top.destroy()
        #print(centimetros)
        A=[]
        B=[]
        for i in range(7):
            u=pixeles[i][0]
            v=pixeles[i][1]
            x=centimetros[i][0]
            y=centimetros[i][1]
            A.append([x,y,0,1,0,0,0,0,-x*u,-y*u,0])
            A.append([0,0,0,0,x,y,0,1,-x*v,-y*v,0])
            B.append(u)
            B.append(v)
        A=np.array(A)
        B=np.array(B)
        #Para matriz mxn ls=(A^+)B
        Api=np.linalg.pinv(A)
        Ls=Api@B
        #print(A)
        #print(B)
        #print(Ls)

if __name__=="__main__":     
    root = tk.Tk()
    app=App(root)
    root.mainloop() 
```
